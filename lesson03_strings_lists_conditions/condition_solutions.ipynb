## 邏輯運算子 (and, or, not)
---

### Q11：完成程式 - and 運算

**解說**：and 要求所有條件都為 True

```python
age = 20
has_id = True

if age >= 18 and has_id:
    print("可以考駕照")
else:
    print("不能考駕照")
```

### Q12：完成程式 - or 運算

**解說**：or 只要有一個條件為 True 即可

```python
is_member = False
amount = 1200

if is_member or amount >= 1000:
    print("有優惠")
else:
    print("無優惠")
```

### Q13：完成程式 - not 運算

**解說**：not 反轉布林值

```python
is_banned = False

if not is_banned:
    print("歡迎使用")
else:
    print("您已被封鎖")
```

### Q14：使用者輸入 - 登入判斷

**解說**：使用 and 檢查兩個條件都滿足

```python
# 測試用：直接賦值
username = "alice"
password = "pass123"

correct_username = "alice"
correct_password = "pass123"

if username == correct_username and password == correct_password:
    print("登入成功")
else:
    print("帳號或密碼錯誤")
```

### Q15：完成程式 - 複雜條件判斷

**解說**：組合 and、or、not 運算子

```python
is_member = True
has_overdue = False
is_vip = False

if (is_member and not has_overdue) or is_vip:
    print("可以借書")
else:
    print("無法借書")
```

### Q19：完成程式 - 簡單巢狀判斷

**解說**：在第一個 if 內再使用 if 進行更細緻的判斷

```python
score = 85

if score >= 60:
    print("及格")
    if score >= 80:
        print("成績優良")
else:
    print("不及格")
```

### Q20：完成程式 - 三角形判斷

**解說**：先判斷能否組成三角形，再判斷類型

```python
a, b, c = 3, 4, 5

if a + b > c and b + c > a and a + c > b:
    print("可以組成三角形")
    
    if a == b == c:
        print("等邊三角形")
    elif a == b or b == c or a == c:
        print("等腰三角形")
    else:
        print("不等邊三角形")
else:
    print("無法組成三角形")
```

### Q21：使用者輸入 - 門票價格

**解說**：根據年齡範圍，再根據學生身分決定價格

```python
# 測試用：直接賦值
age = 20
is_student = True

if age < 6:
    price = 0
    print("免費")
elif age < 18:
    if is_student:
        price = 100
    else:
        price = 150
    print(f"票價: {price} 元")
else:
    if is_student:
        price = 200
    else:
        price = 300
    print(f"票價: {price} 元")
```

### Q22：使用者輸入 - 會員折扣計算

**解說**：根據會員等級計算不同折扣

```python
# 測試用：直接賦值
amount = 1500
member_level = "gold"

if member_level == "none":
    if amount >= 1000:
        final_price = amount * 0.95
    else:
        final_price = amount
elif member_level == "silver":
    final_price = amount * 0.9
elif member_level == "gold":
    final_price = amount * 0.85
elif member_level == "platinum":
    final_price = amount * 0.8
else:
    final_price = amount

print(f"原價: {amount} 元")
print(f"折扣後: {final_price} 元")
```

### Q23：使用者輸入 - 運費計算

**解說**：根據地區和重量計算不同運費

```python
# 測試用：直接賦值
weight = 2.5
region = "north"

if region == "north":
    if weight <= 2:
        shipping = 60
    else:
        shipping = weight * 30
elif region == "central":
    if weight <= 2:
        shipping = 70
    else:
        shipping = weight * 35
elif region == "south":
    if weight <= 2:
        shipping = 80
    else:
        shipping = weight * 40
elif region == "east":
    if weight <= 2:
        shipping = 100
    else:
        shipping = weight * 50
else:
    shipping = 0
    print("地區錯誤")

if shipping > 0:
    print(f"運費: {shipping} 元")
```

### Q25：使用者輸入 - 密碼強度檢查

**解說**：檢查密碼長度和內容組成

```python
# 測試用：直接賦值
password = "Python123"

has_digit = any(c.isdigit() for c in password)
has_alpha = any(c.isalpha() for c in password)

if len(password) < 6:
    print("密碼強度: 弱")
elif len(password) >= 8 and has_digit and has_alpha:
    print("密碼強度: 強")
elif len(password) >= 6 and (has_digit or has_alpha):
    print("密碼強度: 中")
else:
    print("密碼強度: 弱")
```

### Q27：完成程式 - 計算折扣價（多條件）

**解說**：根據數量決定折扣率

```python
quantity = 15
unit_price = 100

if quantity >= 20:
    total = quantity * unit_price * 0.8
    print(f"數量: {quantity}, 折扣: 8 折")
elif quantity >= 10:
    total = quantity * unit_price * 0.9
    print(f"數量: {quantity}, 折扣: 9 折")
else:
    total = quantity * unit_price
    print(f"數量: {quantity}, 無折扣")

print(f"總價: {total} 元")
```

### Q28：使用者輸入 - 飲料點餐

**解說**：使用巢狀判斷處理飲料類型和尺寸

```python
# 測試用：直接賦值
drink = "coffee"
size = "L"
add_ice = True

if drink == "coffee":
    if size == "S":
        price = 50
    elif size == "M":
        price = 60
    else:
        price = 70
elif drink == "tea":
    if size == "S":
        price = 30
    elif size == "M":
        price = 40
    else:
        price = 50
elif drink == "juice":
    if size == "S":
        price = 60
    elif size == "M":
        price = 70
    else:
        price = 80
else:
    price = 0
    print("飲料選擇錯誤")

if price > 0:
    ice_text = "加冰" if add_ice else "去冰"
    print(f"{drink.upper()} {size} {ice_text}")
    print(f"價格: {price} 元")
```

### Q29：完成程式 - 星期判斷

**解說**：根據數字判斷星期並分類

```python
day_number = 3

if day_number == 1:
    day = "星期一"
elif day_number == 2:
    day = "星期二"
elif day_number == 3:
    day = "星期三"
elif day_number == 4:
    day = "星期四"
elif day_number == 5:
    day = "星期五"
elif day_number == 6:
    day = "星期六"
elif day_number == 7:
    day = "星期日"
else:
    day = "錯誤"

if day != "錯誤":
    print(f"今天是 {day}")
    if 1 <= day_number <= 5:
        print("工作日")
    else:
        print("週末")
else:
    print("日期數字錯誤")
```

### Q30：使用者輸入 - 學生成績系統

**解說**：綜合應用所有條件判斷技巧

```python
# 測試用：直接賦值
score = 75

if score < 0 or score > 100:
    print("分數超出範圍 (0-100)")
else:
    # 判斷等第
    if score >= 90:
        grade = "A"
    elif score >= 80:
        grade = "B"
    elif score >= 70:
        grade = "C"
    elif score >= 60:
        grade = "D"
    else:
        grade = "F"
    
    print(f"成績: {score} 分")
    print(f"等第: {grade}")
    
    # 判斷及格與否
    if score >= 60:
        print("及格")
    else:
        print("不及格")
        
        # 判斷補考或重修
        if 40 <= score < 60:
            print("可以參加補考")
        else:
            print("需要重修")
```
