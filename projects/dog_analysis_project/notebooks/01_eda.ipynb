## ğŸ§¹ Step 3: Clean Data

ä½¿ç”¨ `dog.clean_dogs_df()` æ¸…ç†è³‡æ–™ï¼Œä¸¦æ¯”è¼ƒæ¸…ç†å‰å¾Œçš„å·®ç•°ã€‚

```python
# Clean data using our module function
df_clean = dog.clean_dogs_df(df_raw)

# Compare data types before and after
print("ğŸ”„ Data Type Comparison:")
print("\nBefore cleaning:")
print(df_raw.dtypes)
print("\n" + "="*50)
print("\nAfter cleaning:")
print(df_clean.dtypes)
```

## ğŸ“Š Step 4: Generate Statistics

ä½¿ç”¨ `dog.stats_summary()` ç”Ÿæˆçµ±è¨ˆæ‘˜è¦ã€‚

```python
# Generate statistics
stats = dog.stats_summary(df_clean)

# Display in a nice format
print("=" * 60)
print("ğŸ“Š  STATISTICS SUMMARY".center(60))
print("=" * 60)

print(f"\nğŸ“ˆ Overall:")
print(f"   â€¢ Total Dogs: {stats['total_dogs']}")
print(f"   â€¢ Adopted: {stats['adopted_count']} ({stats['adoption_rate']:.1f}%)")
print(f"   â€¢ Vaccinated: {stats['vaccinated_count']}")

print(f"\nğŸ• Physical Stats:")
print(f"   â€¢ Average Age: {stats['avg_age']:.1f} years")
print(f"   â€¢ Average Weight: {stats['avg_weight']:.1f} kg")
print(f"   â€¢ Most Common Breed: {stats['most_common_breed']}")

print(f"\nğŸ™ï¸  By City:")
for city, count in sorted(stats['cities'].items(), key=lambda x: x[1], reverse=True):
    percentage = (count / stats['total_dogs']) * 100
    print(f"   â€¢ {city}: {count} dogs ({percentage:.1f}%)")

print(f"\nğŸ¾ By Breed:")
for breed, count in list(sorted(stats['breeds'].items(), key=lambda x: x[1], reverse=True))[:5]:
    percentage = (count / stats['total_dogs']) * 100
    print(f"   â€¢ {breed}: {count} dogs ({percentage:.1f}%)")
```

## ğŸ” Step 5: Filter and Sort Data

å±•ç¤ºå¦‚ä½•ä½¿ç”¨ `dog.filter_dogs()` å’Œ `dog.sort_dogs()` é€²è¡Œè³‡æ–™ç¯©é¸èˆ‡æ’åºã€‚

```python
# Example 1: Filter by city
print("ğŸ” Example 1: Dogs in Taipei (å°åŒ—)")
taipei_dogs = dog.filter_dogs(df_clean, city="å°åŒ—")
print(taipei_dogs[['name', 'breed', 'age', 'adopted']])
```

```python
# Example 2: Filter by age range
print("ğŸ” Example 2: Young dogs (1-3 years old)")
young_dogs = dog.filter_dogs(df_clean, min_age=1, max_age=3)
print(young_dogs[['name', 'breed', 'age', 'city']])
```

```python
# Example 3: Filter adopted dogs and sort by weight
print("ğŸ” Example 3: Adopted dogs, sorted by weight (descending)")
adopted_dogs = dog.filter_dogs(df_clean, adopted=True)
adopted_sorted = dog.sort_dogs(adopted_dogs, by="weight_kg", ascending=False)
print(adopted_sorted[['name', 'breed', 'age', 'weight_kg', 'city']])
```

```python
# Example 4: Complex filter - Unadopted dogs in Taipei, age 2-5
print("ğŸ” Example 4: Complex filter")
result = dog.filter_dogs(df_clean, city="å°åŒ—", min_age=2, max_age=5, adopted=False)
result_sorted = dog.sort_dogs(result, by="age")
print(result_sorted[['name', 'breed', 'age', 'weight_kg', 'vaccinated']])
```

## ğŸ“ˆ Step 6: Visualizations

ä½¿ç”¨ `dog.plot_top_breeds()` ç”Ÿæˆåœ–è¡¨ï¼Œä¹Ÿå¯ä»¥é¸æ“‡æ˜¯å¦å­˜æª”ã€‚

```python
# Generate and save chart
output_path = project_root / "output" / "top_breeds_from_notebook.png"
print(f"ğŸ’¾ Saving chart to: {output_path}")
dog.plot_top_breeds(df_clean, top_n=5, save_path=output_path)
print("âœ… Chart saved!")
```

```python
# Create additional visualizations
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# 1. Age distribution
axes[0, 0].hist(df_clean['age'], bins=10, color='skyblue', edgecolor='black')
axes[0, 0].set_title('Age Distribution', fontweight='bold')
axes[0, 0].set_xlabel('Age (years)')
axes[0, 0].set_ylabel('Frequency')

# 2. Weight distribution
axes[0, 1].hist(df_clean['weight_kg'], bins=15, color='lightcoral', edgecolor='black')
axes[0, 1].set_title('Weight Distribution', fontweight='bold')
axes[0, 1].set_xlabel('Weight (kg)')
axes[0, 1].set_ylabel('Frequency')

# 3. Adoption status
adoption_counts = df_clean['adopted'].value_counts()
axes[1, 0].bar(['Not Adopted', 'Adopted'], adoption_counts.values, color=['salmon', 'lightgreen'])
axes[1, 0].set_title('Adoption Status', fontweight='bold')
axes[1, 0].set_ylabel('Count')

# 4. City distribution
city_counts = df_clean['city'].value_counts()
axes[1, 1].barh(city_counts.index, city_counts.values, color='plum')
axes[1, 1].set_title('Dogs by City', fontweight='bold')
axes[1, 1].set_xlabel('Count')

plt.tight_layout()
plt.show()

print("ğŸ“Š All visualizations generated!")
```
