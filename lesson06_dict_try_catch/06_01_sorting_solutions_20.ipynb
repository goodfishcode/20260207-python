## 解答 14

### 解題觀念

- 定義函式計算字串中的母音（a, e, i, o, u）數量
- 使用 for 迴圈檢查每個字母是否為母音
- 依照母音數量由少到多排序

```python
words = ['python', 'code', 'awesome', 'fun', 'learn']

def count_vowels(s):
    vowels = "aeiou"
    count = 0
    for letter in s:
        if letter.lower() in vowels:
            count += 1
    return count

result = sorted(words, key=count_vowels)
print(result)

# Output: ['python', 'fun', 'code', 'learn', 'awesome']
```

## 解答 15

### 解題觀念

- Python 的排序是**穩定排序**（stable sort）
- 當兩個元素的 key 值相同時，會保持原本的相對順序
- 例如：`dog`, `cat`, `ant`, `bee`, `fox` 長度都是 3，排序後順序不變
- 這個特性讓排序結果可預測且一致

```python
words = ['dog', 'cat', 'elephant', 'ant', 'bee', 'fox']

result = sorted(words, key=len)
print(result)
print("\n觀察：長度為 3 的字串順序維持原本的 dog → cat → ant → bee → fox")

# Output: ['dog', 'cat', 'ant', 'bee', 'fox', 'elephant']
# 
# 觀察：長度為 3 的字串順序維持原本的 dog → cat → ant → bee → fox
```

## 解答 17

### 解題觀念

- 定義函式或使用 lambda 都可以作為 key

```python
students = [('Alice', 85), ('Bob', 92), ('Charlie', 78), ('David', 95)]

# 方法 1：定義函式
def get_second(item):
    return item[1]

result1 = sorted(students, key=get_second)

# 方法 2：使用 lambda（推薦）
result2 = sorted(students, key=lambda x: x[1])

print(result2)

# Output: [('Charlie', 78), ('Alice', 85), ('Bob', 92), ('David', 95)]
```

## 解答 20

### 解題觀念

- 依平方值排序後，**最後一個元素**是平方值最大的數字
- `sorted_nums[-1]` 使用負數索引取得最後一個元素
