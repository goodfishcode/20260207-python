## ğŸ¯ key è‡ªè¨‚æ’åºè¦å‰‡

æœ‰æ™‚å€™é è¨­çš„æ’åºè¦å‰‡ä¸ç¬¦åˆéœ€æ±‚ï¼Œæˆ‘å€‘å¯ä»¥ç”¨ `key` åƒæ•¸è‡ªè¨‚æ’åºé‚è¼¯ã€‚

**èªæ³•**ï¼š`sorted(list, key=function)`

`key` åƒæ•¸æ¥å—ä¸€å€‹å‡½å¼ï¼ŒPython æœƒç”¨é€™å€‹å‡½å¼çš„å›å‚³å€¼ä¾†æ±ºå®šæ’åºé †åºã€‚

### åŸºæœ¬æ¦‚å¿µ


- Python æœƒå°ä¸²åˆ—ä¸­çš„æ¯å€‹å…ƒç´ åŸ·è¡Œã€ŒæŸå€‹å‡½å¼ã€
- ç”¨å‡½å¼çš„å›å‚³å€¼ä¾†æ’åº
- å¸¸ç”¨çš„å…§å»ºå‡½å¼ï¼š`len`ã€`str.lower`ã€`abs` ç­‰

````python
# åŸºæœ¬ç¯„ä¾‹ï¼šä¾çµ•å°å€¼æ’åº
num = [-5, 2, -8, 1, 9]

print("ä¸€èˆ¬æ’åºï¼š", sorted(num))
print("ä¾çµ•å°å€¼æ’åºï¼š", sorted(num, key=abs))

# é æœŸè¼¸å‡ºï¼š
# ä¸€èˆ¬æ’åºï¼š [-8, -5, 1, 2, 9]
# ä¾çµ•å°å€¼æ’åºï¼š [1, 2, -5, -8, 9]
# èªªæ˜ï¼šabs(1)=1 < abs(2)=2 < abs(-5)=5 < abs(-8)=8 < abs(9)=9
````

## ğŸ“ key=len ä¾é•·åº¦æ’åº

`len` æ˜¯ Python å…§å»ºå‡½å¼ï¼Œå¯ä»¥è¨ˆç®—å­—ä¸²é•·åº¦ã€‚

### ç¯„ä¾‹ï¼šä¾å­—ä¸²é•·åº¦æ’åº

````python
# ä¾å­—ä¸²é•·åº¦æ’åºï¼ˆç”±çŸ­åˆ°é•·ï¼‰
words = ["python", "is", "awesome", "fun"]

result = sorted(words, key=len)
print("ä¾é•·åº¦æ’åºï¼š", result)

# é æœŸè¼¸å‡ºï¼š
# ä¾é•·åº¦æ’åºï¼š ['is', 'fun', 'python', 'awesome']
# èªªæ˜ï¼šis(2) < fun(3) < python(6) < awesome(7)
````

## ğŸ”¤ key=è‡ªè¨‚å‡½å¼ï¼ˆç¬¬äºŒå€‹å­—æ¯ï¼‰

å¦‚æœå…§å»ºå‡½å¼ç„¡æ³•æ»¿è¶³éœ€æ±‚ï¼Œæˆ‘å€‘å¯ä»¥è‡ªå·±å¯«ä¸€å€‹å‡½å¼ã€‚

### ç¯„ä¾‹ï¼šä¾ç¬¬äºŒå€‹å­—æ¯æ’åº

````python
# å®šç¾©ä¸€å€‹å‡½å¼ï¼šå›å‚³ç¬¬äºŒå€‹å­—æ¯
def second_letter(word):
    return word[1]

# ä¾ç¬¬äºŒå€‹å­—æ¯æ’åº
words = ["apple", "banana", "cherry", "date"]
result = sorted(words, key=second_letter)

print("åŸå§‹ï¼š", words)
print("ä¾ç¬¬äºŒå€‹å­—æ¯æ’åºï¼š", result)

# é æœŸè¼¸å‡ºï¼š
# åŸå§‹ï¼š ['apple', 'banana', 'cherry', 'date']
# ä¾ç¬¬äºŒå€‹å­—æ¯æ’åºï¼š ['banana', 'date', 'cherry', 'apple']
# èªªæ˜ï¼šbanana(a) < date(a) < cherry(h) < apple(p)
````

````python
# é©—è­‰ç¬¬äºŒå€‹å­—æ¯
words = ["apple", "banana", "cherry", "date"]

for word in words:
    print(f"{word}: ç¬¬äºŒå€‹å­—æ¯æ˜¯ '{word[1]}'")

# é æœŸè¼¸å‡ºï¼š
# apple: ç¬¬äºŒå€‹å­—æ¯æ˜¯ 'p'
# banana: ç¬¬äºŒå€‹å­—æ¯æ˜¯ 'a'
# cherry: ç¬¬äºŒå€‹å­—æ¯æ˜¯ 'h'
# date: ç¬¬äºŒå€‹å­—æ¯æ˜¯ 'a'
````

### ç¯„ä¾‹ 2ï¼šé©—è­‰ç¬¬äºŒå€‹å­—æ¯

### ğŸ’¡ ç†è§£ key çš„é‹ä½œæ–¹å¼

ç•¶æˆ‘å€‘å¯« `sorted(words, key=second_letter)` æ™‚ï¼ŒPython æœƒï¼š

1. å°æ¯å€‹å–®å­—åŸ·è¡Œ `second_letter()` å‡½å¼
   - `second_letter("apple")` â†’ `"p"`
   - `second_letter("banana")` â†’ `"a"`
   - `second_letter("cherry")` â†’ `"h"`
   - `second_letter("date")` â†’ `"a"`

2. ä¾ç…§å›å‚³å€¼æ’åºï¼š`"a" < "a" < "h" < "p"`

3. å›å‚³æ’åºå¾Œçš„åŸå§‹å–®å­—

## ğŸµ key=è‡ªè¨‚å‡½å¼ï¼ˆæ¯éŸ³æ•¸é‡ï¼‰

å†ä¾†ä¸€å€‹æ›´è¤‡é›œçš„ä¾‹å­ï¼šä¾æ¯éŸ³ï¼ˆa, e, i, o, uï¼‰æ•¸é‡æ’åºã€‚

### ç¯„ä¾‹ï¼šè¨ˆç®—æ¯éŸ³æ•¸é‡ä¸¦æ’åº

````python
# å®šç¾©å‡½å¼ï¼šè¨ˆç®—æ¯éŸ³æ•¸é‡
def count_vowels(word):
    vowels = "aeiou"
    count = 0
    for letter in word:
        if letter.lower() in vowels:
            count += 1
    return count

# ä¾æ¯éŸ³æ•¸é‡æ’åº
words = ["python", "hello", "world", "awesome"]

result = sorted(words, key=count_vowels)
print("åŸå§‹ï¼š", words)
print("ä¾æ¯éŸ³æ•¸é‡æ’åºï¼š", result)

# é æœŸè¼¸å‡ºï¼š
# åŸå§‹ï¼š ['python', 'hello', 'world', 'awesome']
# ä¾æ¯éŸ³æ•¸é‡æ’åºï¼š ['python', 'world', 'hello', 'awesome']
# èªªæ˜ï¼špython(1) < world(1) < hello(2) < awesome(4)
````

### é©—è­‰ä¸€ä¸‹

````python
# é©—è­‰æ¯å€‹å–®å­—çš„æ¯éŸ³æ•¸é‡
words = ["python", "hello", "world", "awesome"]

for word in words:
    print(f"{word}: {count_vowels(word)} å€‹æ¯éŸ³")

# é æœŸè¼¸å‡ºï¼š
# python: 1 å€‹æ¯éŸ³
# hello: 2 å€‹æ¯éŸ³
# world: 1 å€‹æ¯éŸ³
# awesome: 4 å€‹æ¯éŸ³
````
