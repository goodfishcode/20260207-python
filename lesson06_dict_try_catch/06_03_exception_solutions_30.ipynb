### é¡Œç›® 13 è§£æï¼šä½¿ç”¨è€…è¼¸å…¥é©—è­‰è¿´åœˆ

**ğŸ“ è§£é¡Œé—œéµ**ï¼š

**ğŸ” åŸ·è¡Œçµæœ**ï¼š
- ç¬¬ 1 æ¬¡è¼¸å…¥ "abc" â†’ ç„¡æ•ˆï¼Œå‰©é¤˜ 2 æ¬¡
- ç¬¬ 2 æ¬¡è¼¸å…¥ "5.5" â†’ ç„¡æ•ˆï¼Œå‰©é¤˜ 1 æ¬¡

```python
# é¡Œç›® 13 å®Œæ•´è§£ç­”
user_inputs = ["abc", "5.5", "42"]
input_index = 0
max_retries = 3
attempts = 0

def mock_input():
    global input_index
    if input_index < len(user_inputs):
        val = user_inputs[input_index]
        input_index += 1
        return val
    return None

print("è«‹è¼¸å…¥ä¸€å€‹æ•´æ•¸ï¼ˆæœ€å¤š 3 æ¬¡æ©Ÿæœƒï¼‰")
while attempts < max_retries:
    try:
        user_input = mock_input()
        if user_input is None:
            break
        num = int(user_input)
        print(f"âœ… æ‚¨è¼¸å…¥çš„æ•¸å­—ï¼š{num}")
        break
    except ValueError:
        attempts += 1
        if attempts < max_retries:
            print(f"âŒ ç„¡æ•ˆè¼¸å…¥ï¼Œè«‹é‡è©¦ï¼ˆå‰©é¤˜ {max_retries - attempts} æ¬¡ï¼‰")
        else:
            print("âŒ å·²é”åˆ°æœ€å¤§é‡è©¦æ¬¡æ•¸")
```

### é¡Œç›® 14 è§£æï¼šå®‰å…¨è§£æè³‡æ–™æª”æ¡ˆ

**ğŸ’¡ æ ¸å¿ƒæ¦‚å¿µ**ï¼š
1. **CSV æ•¸æ“šè§£æ**ï¼šä½¿ç”¨ `split()` åˆ†å‰²å­—ç¬¦ä¸²
2. **æ•¸æ“šé©—è­‰**ï¼šæª¢æŸ¥æ¬„ä½æ•¸é‡å’Œé¡å‹
3. **è‡ªå®šç¾© ValueError**ï¼šæ‰‹å‹•æ‹‹å‡ºç•°å¸¸

**ğŸ“ è§£é¡Œé—œéµ**ï¼š
- ä½¿ç”¨ `split(",")` åˆ†å‰² CSV è¡Œ
- æª¢æŸ¥ `len(parts) != 3` æ‹‹å‡º ValueError
- `int()` è½‰æ›å¤±æ•—è‡ªå‹•æ‹‹å‡º ValueError
- æ‰€æœ‰ ValueError çµ±ä¸€è™•ç†ï¼Œè·³éç„¡æ•ˆè¡Œ

**ğŸ” åŸ·è¡Œçµæœ**ï¼š
- "Alice,25,50000" â†’ æˆåŠŸè§£æ
- "Bob,thirty,60000" â†’ ValueErrorï¼ˆå¹´é½¡ç„¡æ³•è½‰æ•´æ•¸ï¼‰
- "Charlie,35" â†’ ValueErrorï¼ˆæ¬„ä½ä¸è¶³ï¼‰
- "Diana,28,70000" â†’ æˆåŠŸè§£æ

```python
# é¡Œç›® 14 å®Œæ•´è§£ç­”

def parse_csv(data):
    """å®‰å…¨è§£æ CSV æ ¼å¼çš„è³‡æ–™"""
    result = []
    for line in data:
        try:
            # åˆ†å‰²è³‡æ–™
            parts = line.split(",")
            
            # æª¢æŸ¥æ¬„ä½æ•¸é‡
            if len(parts) != 3:
                raise ValueError("æ¬„ä½æ•¸é‡ä¸æ­£ç¢º")
            
            # è§£æå„æ¬„ä½
            name = parts[0]
            age = int(parts[1])  # å¯èƒ½æ‹‹å‡º ValueError
            salary = int(parts[2])  # å¯èƒ½æ‹‹å‡º ValueError
            
            # åŠ å…¥çµæœ
            result.append({"name": name, "age": age, "salary": salary})
            
        except ValueError as e:
            print(f"è­¦å‘Š: ç„¡æ³•è§£æ '{line}' - {e}")
    
    return result

# æ¸¬è©¦è³‡æ–™
csv_data = [
    "Alice,25,50000",
    "Bob,thirty,60000",    # å¹´é½¡ç„¡æ³•è½‰æ•´æ•¸
    "Charlie,35",          # æ¬„ä½ä¸è¶³
    "Diana,28,70000"
]

print("è§£æçµæœï¼š")
parsed_data = parse_csv(csv_data)
for person in parsed_data:
    print(person)
```

### é¡Œç›® 15 è§£æï¼šè¨ˆç®—æ©Ÿæ‡‰ç”¨

**ğŸ’¡ æ ¸å¿ƒæ¦‚å¿µ**ï¼š
1. **å­—å…¸é‹ç®—ç¬¦æ˜ å°„**ï¼šä½¿ç”¨ dict å„²å­˜é‹ç®—å‡½æ•¸
2. **KeyError è™•ç†**ï¼šæ•ç²ä¸æ”¯æ´çš„é‹ç®—ç¬¦
3. **ZeroDivisionError è™•ç†**ï¼šæ•ç²é™¤é›¶éŒ¯èª¤
4. **é‹ç®—ç¬¦æ¨¡å¼åŒ¹é…**ï¼šä½¿ç”¨å­—å…¸å–ä»£ if-elif

**ğŸ“ è§£é¡Œé—œéµ**ï¼š
- ä½¿ç”¨ `operations = {"+": lambda...}` æ˜ å°„é‹ç®—ç¬¦
- `operations[operator]` æŸ¥æ‰¾é‹ç®—ï¼Œä¸å­˜åœ¨æ™‚æ‹‹å‡º KeyError
- é™¤æ³•é‹ç®— `b = 0` æ™‚æ‹‹å‡º ZeroDivisionError
- åˆ†åˆ¥è™•ç†å…©ç¨®ç•°å¸¸ï¼Œçµ¦å‡ºæ˜ç¢ºéŒ¯èª¤è¨Šæ¯

**ğŸ” åŸ·è¡Œçµæœ**ï¼š
- calculate(10, 5, "+") â†’ 15
- calculate(10, 5, "/") â†’ 2.0
- calculate(10, 0, "/") â†’ ZeroDivisionErrorï¼ˆé™¤ä»¥é›¶ï¼‰
- calculate(10, 5, "%") â†’ KeyErrorï¼ˆä¸æ”¯æ´çš„é‹ç®—ç¬¦ï¼‰

```python
# é¡Œç›® 15 å®Œæ•´è§£ç­”

def calculate(a, b, operator):
    """åŸ·è¡ŒåŸºæœ¬é‹ç®—ä¸¦è™•ç†ç•°å¸¸"""
    # å®šç¾©é‹ç®—ç¬¦å­—å…¸
    operations = {
        "+": lambda x, y: x + y,
        "-": lambda x, y: x - y,
        "*": lambda x, y: x * y,
        "/": lambda x, y: x / y
    }
    
    try:
        # æŸ¥æ‰¾ä¸¦åŸ·è¡Œé‹ç®—
        operation = operations[operator]  # å¯èƒ½æ‹‹å‡º KeyError
        result = operation(a, b)          # å¯èƒ½æ‹‹å‡º ZeroDivisionError
        return result
        
    except KeyError:
        print(f"éŒ¯èª¤: ä¸æ”¯æ´çš„é‹ç®—ç¬¦ '{operator}'")
        print("æ”¯æ´çš„é‹ç®—ç¬¦: +, -, *, /")
        
    except ZeroDivisionError:
        print("éŒ¯èª¤: é™¤æ•¸ä¸èƒ½ç‚ºé›¶")

# æ¸¬è©¦
print("10 + 5 =", calculate(10, 5, "+"))
print("10 / 5 =", calculate(10, 5, "/"))
print("\næ¸¬è©¦ç•°å¸¸æƒ…æ³ï¼š")
calculate(10, 0, "/")  # ZeroDivisionError
calculate(10, 5, "%")  # KeyError
```

```python
# é¡Œç›® 21 å®Œæ•´è§£ç­”

balance = 1000

def process_transaction(amount, balance):
    """è™•ç†äº¤æ˜“"""
    # è½‰æ›é¡å‹
    amount = float(amount)
    
    # é©—è­‰é‡‘é¡
    if amount <= 0:
        raise ValueError("é‡‘é¡å¿…é ˆå¤§æ–¼ 0")
    
    # æª¢æŸ¥é¤˜é¡
    if amount > balance:
        raise RuntimeError("é¤˜é¡ä¸è¶³")
    
    return balance - amount

# æ¸¬è©¦è³‡æ–™
transactions = [500, -100, 2000, "invalid"]

print("äº¤æ˜“é‡‘é¡é©—è­‰ï¼š")
current_balance = balance
for amount in transactions:
    print(f"\näº¤æ˜“é‡‘é¡ï¼š{amount}")
    try:
        current_balance = process_transaction(amount, current_balance)
        print(f"âœ… äº¤æ˜“æˆåŠŸï¼š{amount} å…ƒï¼Œå‰©é¤˜ {current_balance} å…ƒ")
    except ValueError:
        print("âŒ äº¤æ˜“å¤±æ•—ï¼šé‡‘é¡ä¸èƒ½ç‚ºè² æ•¸æˆ–é›¶")
    except RuntimeError:
        print("âŒ äº¤æ˜“å¤±æ•—ï¼šé¤˜é¡ä¸è¶³")
    except TypeError:
        print("âŒ äº¤æ˜“å¤±æ•—ï¼šé‡‘é¡æ ¼å¼éŒ¯èª¤")
```

- æª¢æŸ¥ `user["locked"]` ç‹€æ…‹ï¼Œé–å®šå‰‡æ‹‹å‡º RuntimeError

```python
# é¡Œç›® 22 å®Œæ•´è§£ç­”

users = {
    "alice": {"password": "123", "locked": False},
    "bob": {"password": "456", "locked": True}
}

def login(username, password):
    """é©—è­‰ç™»å…¥"""
    # æª¢æŸ¥å¸³è™Ÿå­˜åœ¨
    user = users[username]
    
    # æª¢æŸ¥å¸³è™Ÿç‹€æ…‹
    if user["locked"]:
        raise RuntimeError("å¸³è™Ÿå·²é–å®š")
    
    # é©—è­‰å¯†ç¢¼
    if user["password"] != password:
        raise ValueError("å¯†ç¢¼éŒ¯èª¤")
    
    return True

# æ¸¬è©¦è³‡æ–™
login_attempts = [
    ("alice", "123"),
    ("alice", "wrong"),
    ("bob", "456"),
    ("charlie", "123")
]

print("ç™»å…¥é©—è­‰ç³»çµ±ï¼š")
for username, password in login_attempts:
    print(f"\nå˜—è©¦ç™»å…¥ï¼š{username}")
    try:
        login(username, password)
        print(f"âœ… ç™»å…¥æˆåŠŸï¼Œæ­¡è¿ {username}")
    except KeyError:
        print("âŒ ç™»å…¥å¤±æ•—ï¼šå¸³è™Ÿä¸å­˜åœ¨")
    except ValueError:
        print("âŒ ç™»å…¥å¤±æ•—ï¼šå¯†ç¢¼éŒ¯èª¤")
    except RuntimeError:
        print("âŒ ç™»å…¥å¤±æ•—ï¼šå¸³è™Ÿå·²é–å®š")
```

```python
# é¡Œç›® 23 å®Œæ•´è§£ç­”

available_files = {"data.txt": "file content"}

def read_file_safe(filename):
    """å®‰å…¨åœ°è®€å–æª”æ¡ˆ"""
    f = None
    try:
        # æª¢æŸ¥æª”æ¡ˆæ˜¯å¦å­˜åœ¨
        if filename not in available_files:
            raise FileNotFoundError(f"æª”æ¡ˆä¸å­˜åœ¨ï¼š{filename}")
        
        # æ¨¡æ“¬é–‹å•Ÿæª”æ¡ˆ
        f = f"<file object: {filename}>"
        print(f"  âœ“ é–‹å•Ÿæª”æ¡ˆï¼š{filename}")
        
        # è®€å–å…§å®¹
        return available_files[filename]
        
    except FileNotFoundError as e:
        print(f"  âŒ {str(e)}")
        return None
        
    finally:
        # ç„¡è«–æˆåŠŸæˆ–å¤±æ•—ï¼Œéƒ½é—œé–‰æª”æ¡ˆ
        if f:
            print(f"  âœ“ é—œé–‰æª”æ¡ˆ")

# æ¸¬è©¦è³‡æ–™
files = ["data.txt", "missing.txt"]

print("ä½¿ç”¨ finally æ¸…ç†è³‡æºï¼š")
for filename in files:
    print(f"\nè®€å–ï¼š{filename}")
    content = read_file_safe(filename)
    if content:
        print(f"âœ… è®€å–æˆåŠŸï¼š{filename}")
```

### é¡Œç›® 27 è§£æï¼šåµŒå¥—ç•°å¸¸è™•ç†

**ğŸ’¡ æ ¸å¿ƒæ¦‚å¿µ**ï¼š

4. **å¤šå±¤éŒ¯èª¤è™•ç†**ï¼šæ¨¡æ“¬å¯¦éš›æ‡‰ç”¨ä¸­çš„è¤‡é›œéŒ¯èª¤æƒ…å¢ƒ

**ğŸ“ è§£é¡Œé—œéµ**ï¼š

- å…§å±¤ tryï¼šè™•ç† ValueErrorï¼ˆè§£æéŒ¯èª¤ï¼‰
- å…§å±¤ç•°å¸¸æ•ç²å¾Œä½¿ç”¨ `raise` é‡æ–°æ‹‹å‡ºåˆ°å¤–å±¤
- finally ç¢ºä¿ç„¡è«–å“ªå±¤å‡ºéŒ¯éƒ½é—œé–‰è³‡æº
- æœ€å¤–å±¤å†æ¬¡æ•ç²è™•ç†æ‰€æœ‰å¯èƒ½çš„ç•°å¸¸

**ğŸ” åŸ·è¡Œçµæœ**ï¼š
- "valid.txt" â†’ âœ“ è®€å– â†’ âœ“ è§£æ â†’ âœ“ é—œé–‰ â†’ âœ… æˆåŠŸè®€å–è³‡æ–™
- "invalid.txt" â†’ âœ“ è®€å– â†’ âŒ è§£æå¤±æ•— â†’ âœ“ é—œé–‰ â†’ âŒ ç„¡æ³•å®Œæˆæ“ä½œ
- "missing.txt" â†’ âŒ æª”æ¡ˆä¸å­˜åœ¨ â†’ âœ“ é—œé–‰ â†’ âŒ ç„¡æ³•å®Œæˆæ“ä½œ

```python
# é¡Œç›® 27 å®Œæ•´è§£ç­”

files_content = {
    "valid.txt": "name:Alice,age:25",
    "invalid.txt": "invalid data"
}

def read_and_parse_file(filename):
    """è®€å–ä¸¦è§£ææª”æ¡ˆ"""
    try:
        # å¤–å±¤ï¼šè™•ç†æª”æ¡ˆéŒ¯èª¤
        if filename not in files_content:
            raise FileNotFoundError(f"æª”æ¡ˆä¸å­˜åœ¨ï¼š{filename}")
        
        content = files_content[filename]
        print(f"  âœ“ è®€å–æª”æ¡ˆ")
        
        try:
            # å…§å±¤ï¼šè™•ç†è§£æéŒ¯èª¤
            parts = content.split(",")
            if len(parts) < 2:
                raise ValueError("è³‡æ–™æ ¼å¼ä¸æ­£ç¢º")
            
            name = parts[0].split(":")[1]
            age = parts[1].split(":")[1]
            print(f"  âœ“ è§£æè³‡æ–™")
            return name, age
            
        except ValueError:
            print(f"  âŒ è§£æå¤±æ•—")
            raise  # é‡æ–°æ‹‹å‡ºç•°å¸¸
            
    except FileNotFoundError as e:
        print(f"  âŒ {str(e)}")
        raise  # é‡æ–°æ‹‹å‡ºç•°å¸¸
        
    finally:
        print(f"  âœ“ é—œé–‰æª”æ¡ˆ")

# æ¸¬è©¦è³‡æ–™
files = ["valid.txt", "invalid.txt", "missing.txt"]

print("åµŒå¥—ç•°å¸¸è™•ç†ï¼š")
for filename in files:
    print(f"\nè®€å–ï¼š{filename}")
    try:
        name, age = read_and_parse_file(filename)
        print(f"âœ… æˆåŠŸè®€å–è³‡æ–™ï¼š{name}, {age}")
    except (FileNotFoundError, ValueError):
        print(f"âŒ ç„¡æ³•å®Œæˆæ“ä½œ")
```

### é¡Œç›® 28 è§£æï¼šè‡ªå®šç¾©ç•°å¸¸

**ğŸ’¡ æ ¸å¿ƒæ¦‚å¿µ**ï¼š

3. **ç•°å¸¸å±¤æ¬¡**ï¼šå€åˆ†ç³»çµ±ç•°å¸¸ï¼ˆValueErrorï¼‰å’Œæ¥­å‹™ç•°å¸¸ï¼ˆAgeErrorï¼‰
4. **ç•°å¸¸è¨­è¨ˆ**ï¼šè®“ç•°å¸¸åç¨±ç›´æ¥åæ˜ å•é¡Œ

**ğŸ“ è§£é¡Œé—œéµ**ï¼š
- å®šç¾© `class AgeError(Exception)`ï¼Œå¯æ·»åŠ  docstring
- ValueErrorï¼šé¡å‹éŒ¯èª¤ï¼ˆå¦‚éæ•´æ•¸ï¼‰
- AgeErrorï¼šæ¥­å‹™è¦å‰‡éŒ¯èª¤ï¼ˆå¹´é½¡ç¯„åœï¼‰
- åˆ†åˆ¥æ•ç²è™•ç†ï¼Œçµ¦å‡ºä¸åŒè¨Šæ¯

**ğŸ” åŸ·è¡Œçµæœ**ï¼š
- Alice (20) â†’ âœ… å­¸ç”Ÿå·²ç™»è¨˜
- Bob (15) â†’ âŒ å¹´é½¡ä¸ç¬¦åˆè¦æ±‚ï¼ˆ< 18ï¼‰
- Charlie (25) â†’ âœ… å­¸ç”Ÿå·²ç™»è¨˜
- Diana (70) â†’ âŒ å¹´é½¡ä¸ç¬¦åˆè¦æ±‚ï¼ˆ> 65ï¼‰

```python
# é¡Œç›® 28 å®Œæ•´è§£ç­”

class AgeError(Exception):
    """å¹´é½¡ä¸ç¬¦åˆè¦æ±‚çš„ç•°å¸¸"""
    pass

def register_student(name, age):
    """è¨»å†Šå­¸ç”Ÿ"""
    # æª¢æŸ¥é¡å‹
    if not isinstance(age, int):
        raise ValueError("å¹´é½¡å¿…é ˆæ˜¯æ•´æ•¸")
    
    # æª¢æŸ¥å¹´é½¡ç¯„åœ
    if age < 18 or age > 65:
        raise AgeError(f"å¹´é½¡ {age} ä¸ç¬¦åˆè¦æ±‚ï¼ˆå¿…é ˆ 18-65 æ­²ï¼‰")
    
    return True

# æ¸¬è©¦è³‡æ–™
students_data = [
    {"name": "Alice", "age": 20},
    {"name": "Bob", "age": 15},
    {"name": "Charlie", "age": 25},
    {"name": "Diana", "age": 70}
]

print("è‡ªå®šç¾©ç•°å¸¸ï¼š")
for student in students_data:
    name = student["name"]
    age = student["age"]
    print(f"\nè¨»å†Šå­¸ç”Ÿï¼š{name}ï¼ˆå¹´é½¡ {age}ï¼‰")
    try:
        register_student(name, age)
        print(f"âœ… å­¸ç”Ÿ {name} å·²ç™»è¨˜")
    except ValueError as e:
        print(f"âŒ è¨»å†Šå¤±æ•—ï¼š{str(e)}")
    except AgeError:
        print(f"âŒ å¹´é½¡ä¸ç¬¦åˆè¦æ±‚")
```

### é¡Œç›® 29 è§£æï¼šå¤šå±¤è³‡æ–™é©—è­‰

**ğŸ’¡ æ ¸å¿ƒæ¦‚å¿µ**ï¼š
1. **åˆ†å±¤é©—è­‰**ï¼šå°‡è¤‡é›œé©—è­‰æ‹†åˆ†ç‚ºå¤šå€‹æ­¥é©Ÿ
2. **çµ±ä¸€ç•°å¸¸é¡å‹**ï¼šå¤šç¨®éŒ¯èª¤çµ±ä¸€ç”¨ ValueError è¡¨é”
3. **ç•°å¸¸è¨Šæ¯**ï¼šç”¨è©³ç´°è¨Šæ¯èªªæ˜å…·é«”éŒ¯èª¤åŸå› 
4. **å®¹éŒ¯è¨­è¨ˆ**ï¼šä½¿ç”¨ `get()` æ–¹æ³•å®‰å…¨å–å¾—å¯é¸æ¬„ä½

**ğŸ“ è§£é¡Œé—œéµ**ï¼š
- ç¬¬ä¸€å±¤ï¼šé©—è­‰å€‹äººè³‡è¨Šï¼ˆname å­˜åœ¨ä¸”æœ‰æ•ˆï¼‰
- ç¬¬äºŒå±¤ï¼šé©—è­‰éƒµç®±ï¼ˆåŒ…å« @ï¼‰
- ç¬¬ä¸‰å±¤ï¼šé©—è­‰é›»è©±ï¼ˆ10 ä½æ•¸å­—ï¼‰
- æ¯å±¤ç”¨ try-except æ•ç² KeyErrorï¼Œè½‰ç‚º ValueError
- æ‰€æœ‰éŒ¯èª¤çµ±ä¸€åœ¨å¤–å±¤è™•ç†

**ğŸ” åŸ·è¡Œçµæœ**ï¼š
- Alice â†’ âœ… ä½¿ç”¨è€…è³‡æ–™æœ‰æ•ˆï¼ˆå…¨éƒ¨é€šéï¼‰
- Bob â†’ âš ï¸ é©—è­‰å¤±æ•—ï¼šéƒµç®±æ ¼å¼ç„¡æ•ˆ

```python
# é¡Œç›® 29 å®Œæ•´è§£ç­”

def validate_user(user):
    """é©—è­‰ä½¿ç”¨è€…è³‡æ–™"""
    # ç¬¬ä¸€å±¤ï¼šé©—è­‰å€‹äººè³‡è¨Š
    try:
        name = user["name"]
        if not isinstance(name, str) or not name:
            raise ValueError("åç¨±ç„¡æ•ˆ")
    except KeyError:
        raise ValueError("ç¼ºå°‘åç¨±æ¬„ä½")
    
    # ç¬¬äºŒå±¤ï¼šé©—è­‰éƒµç®±
    try:
        email = user["email"]
        if "@" not in email:
            raise ValueError("éƒµç®±æ ¼å¼ç„¡æ•ˆ")
    except KeyError:
        raise ValueError("ç¼ºå°‘éƒµç®±æ¬„ä½")
    
    # ç¬¬ä¸‰å±¤ï¼šé©—è­‰é›»è©±
    try:
        phone = user["phone"]
        if not phone.isdigit() or len(phone) != 10:
            raise ValueError("é›»è©±è™Ÿç¢¼ç„¡æ•ˆ")
    except (KeyError, AttributeError):
        raise ValueError("é›»è©±è™Ÿç¢¼ç„¡æ•ˆ")
    
    return True

# æ¸¬è©¦è³‡æ–™
users = [
    {"name": "Alice", "email": "alice@example.com", "phone": "1234567890"},
    {"name": "Bob", "email": "invalid", "phone": "1234567890"},
    {"name": "Charlie", "email": "charlie@example.com", "phone": "invalid"}
]

print("å¤šå±¤è³‡æ–™é©—è­‰ï¼š")
for user in users:
    print(f"\né©—è­‰ä½¿ç”¨è€…ï¼š{user.get('name', 'æœªçŸ¥')}")
    try:
        validate_user(user)
        print(f"âœ… ä½¿ç”¨è€…è³‡æ–™æœ‰æ•ˆ")
    except ValueError as e:
        print(f"âš ï¸ é©—è­‰å¤±æ•—ï¼š{str(e)}")
```

### é¡Œç›® 30 è§£æï¼šç•°å¸¸é‡æ–°æ‹‹å‡º

**ğŸ’¡ æ ¸å¿ƒæ¦‚å¿µ**ï¼š
1. **ç•°å¸¸é‡æ–°æ‹‹å‡º**ï¼šä½¿ç”¨ `raise` ä¸å¸¶åƒæ•¸é‡æ–°æ‹‹å‡ºç•¶å‰ç•°å¸¸
2. **æ—¥èªŒè¨˜éŒ„**ï¼šåœ¨é‡æ–°æ‹‹å‡ºå‰è¨˜éŒ„ç•°å¸¸è¨Šæ¯
3. **åˆ†å±¤è™•ç†**ï¼šå…§å±¤è¨˜éŒ„ï¼Œå¤–å±¤çµ±ä¸€è™•ç†
4. **çµ±è¨ˆåˆ†æ**ï¼šè¿½è¹¤æˆåŠŸå’Œå¤±æ•—çš„æ“ä½œ

**ğŸ“ è§£é¡Œé—œéµ**ï¼š
- å…§å±¤ try-exceptï¼šæ•ç²ç•°å¸¸ï¼Œè¨˜éŒ„æ—¥èªŒï¼Œä½¿ç”¨ `raise` é‡æ‹‹
- å¤–å±¤ try-exceptï¼šæ•ç²é‡æ‹‹çš„ç•°å¸¸ï¼Œé€²è¡Œçµ±ä¸€è™•ç†
- ä½¿ç”¨åˆ—è¡¨ `failed_operations` è¿½è¹¤å¤±æ•—æ“ä½œ
- æœ€å¾Œçµ±è¨ˆæˆåŠŸ/å¤±æ•—æ•¸é‡

**ğŸ” åŸ·è¡Œçµæœ**ï¼š
- divide(10, 2) â†’ âœ… çµæœï¼š5.0
- divide(10, 0) â†’ âš ï¸ å…§å±¤ç•°å¸¸ï¼šZeroDivisionError â†’ âŒ æ“ä½œå¤±æ•—
- access([1,2], 5) â†’ âš ï¸ å…§å±¤ç•°å¸¸ï¼šIndexError â†’ âŒ æ“ä½œå¤±æ•—
- multiply(5, 3) â†’ âœ… çµæœï¼š15
- çµ±è¨ˆï¼šæˆåŠŸ 2ï¼Œå¤±æ•— 2

```python
# é¡Œç›® 30 å®Œæ•´è§£ç­”

def perform_operation(op, a, b):
    """åŸ·è¡Œæ“ä½œ"""
    try:
        if op == "divide":
            return a / b
        elif op == "access":
            return a[b]
        elif op == "multiply":
            return a * b
        else:
            raise ValueError(f"æœªçŸ¥æ“ä½œï¼š{op}")
            
    except (ZeroDivisionError, IndexError, ValueError) as e:
        # è¨˜éŒ„æ—¥èªŒ
        print(f"  âš ï¸ å…§å±¤ç•°å¸¸ï¼š{type(e).__name__}")
        # é‡æ–°æ‹‹å‡ºç•°å¸¸
        raise

# æ¸¬è©¦è³‡æ–™
operations = [
    ("divide", 10, 2),
    ("divide", 10, 0),
    ("access", [1, 2], 5),
    ("multiply", 5, 3)
]

print("ç•°å¸¸é‡æ–°æ‹‹å‡ºï¼š")
failed_operations = []

for op, a, b in operations:
    print(f"\nåŸ·è¡Œï¼š{op}({a}, {b})")
    try:
        result = perform_operation(op, a, b)
        print(f"âœ… çµæœï¼š{result}")
    except Exception as e:
        print(f"âŒ æ“ä½œå¤±æ•—ï¼š{type(e).__name__}")
        failed_operations.append((op, str(e)))

print(f"\nğŸ“Š çµ±è¨ˆ")
print(f"æˆåŠŸæ“ä½œï¼š{len(operations) - len(failed_operations)}")
print(f"å¤±æ•—æ“ä½œï¼š{len(failed_operations)}")
print(f"âœ… æ‰€æœ‰æ“ä½œå®Œæˆ")
```
