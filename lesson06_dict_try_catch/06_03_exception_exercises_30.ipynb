```python
# é¡Œç›® 15ï¼šè¨ˆç®—æ©Ÿæ‡‰ç”¨
test_cases = [
    ("10", "+", "5"),
    ("10", "-", "3"),
    ("10", "*", "2"),
    ("10", "/", "2"),
    ("10", "/", "0"),
    ("abc", "+", "5"),
    ("10", "%", "5")
]

def calculate(a, op, b):
    """åŸ·è¡Œè¨ˆç®—"""
    a_num = int(a)
    b_num = int(b)
    
    ops = {
        "+": a_num + b_num,
        "-": a_num - b_num,
        "*": a_num * b_num,
        "/": a_num / b_num
    }
    return ops[op]

print("è¨ˆç®—æ©Ÿæ‡‰ç”¨ï¼š")
for a, op, b in test_cases:
    print(f"\nè¨ˆç®—ï¼š{a} {op} {b}")
    try:
        result = calculate(a, op, b)
        print(f"âœ… {a} {op} {b} = {result}")
    except ValueError:
        print("âŒ è¼¸å…¥ä¸æ˜¯æ•¸å­—")
    except ZeroDivisionError:
        print("âŒ é™¤æ•¸ä¸å¯ç‚ºé›¶")
    except KeyError:
        print(f"âŒ ä¸æ”¯æŒçš„æ“ä½œç¬¦ï¼š{op}")
    except TypeError:
        print("âŒ é¡å‹éŒ¯èª¤")
```
