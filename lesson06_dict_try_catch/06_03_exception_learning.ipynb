## 什麼是例外（Exception）

### ✅ 定義
**例外（Exception）** 是指程式在**執行期間**發生的錯誤。

### 🔍 例外的特性
- 程式語法正確，但執行時遇到問題
- 如果**沒有處理**，程式會**立即中斷**
- 使用者會看到紅色的錯誤訊息

### 📝 常見的例外類型
- **ValueError**：值錯誤（例如：將「五」轉成整數）
- **ZeroDivisionError**：除以零
- **NameError**：變數名稱錯誤（變數不存在）
- **TypeError**：類型錯誤（例如：對數字使用 len()）
- **KeyError**：字典找不到 key
- **IndexError**：list 索引超出範圍

### 💡 為什麼需要了解例外
程式在真實世界中，經常會遇到：
- 使用者輸入錯誤的資料
- 檔案不存在
- 網路連線失敗
- 資料格式不符

**學會處理例外，才能寫出穩定的程式！**

## ⚠️ 示範：沒有例外處理會怎樣

下面的程式碼會**故意產生錯誤**，但我們用 `try/except` 把它接住，這樣 notebook 就不會真的中斷。

**重點**：如果拿掉 `try/except`，程式會直接炸掉，後面的程式碼都不會執行！

````python
# 示範：沒有例外處理會怎樣（但我們用 try/except 接住它）

print("=== 示範開始 ===")

try:
    # ⚠️ 這行會出錯：無法將中文數字轉成整數
    result = int("五")
    print(f"結果：{result}")
except:
    print("❌ 發生錯誤！但我們用 try/except 接住了")
    print("錯誤類型：無法將「五」轉換成整數")

print("✅ 程式繼續執行，沒有中斷！")
print("=== 示範結束 ===")

# 📌 重要說明：
# 如果把上面的 try/except 拿掉，只留下：
#   result = int("五")
# 程式會直接炸掉，後面的 print 都不會執行！
````

## 為什麼要例外處理

### ✅ 四大理由

#### 1. 使用者輸入不一定正確
- 使用者可能輸入「五」而不是「5」
- 使用者可能輸入「abc」而不是數字

#### 2. 讓程式更穩定
- 遇到錯誤不會直接崩潰
- 可以記錄錯誤、通知管理員
- 可以嘗試重試或使用備援方案

#### 3. 給使用者友善的提示
```
❌ 不好的錯誤訊息：
ValueError: invalid literal for int() with base 10: '五'

✅ 友善的提示：
請輸入阿拉伯數字（例如：5）
```

#### 4. 讓後續程式繼續執行
- 一個錯誤不應該影響整個系統
- 可以跳過錯誤的資料，繼續處理其他資料
- 例如：處理 100 筆資料，1 筆錯誤不應該讓其他 99 筆都無法處理

### 💡 結論
**例外處理是專業程式設計的必備技能！**

## try / except 基本語法

### 📝 語法格式

```python
try:
    # 可能會出錯的程式碼
    # 如果成功執行，就跳過 except
except:
    # 發生錯誤時執行這裡
    # 處理錯誤、顯示訊息、記錄日誌等
```

### 🔄 執行流程

1. **Python 先執行 `try` 區塊**
   - 如果**沒有錯誤**：執行完 try 後，**跳過 except**，繼續執行後面的程式碼
   - 如果**有錯誤**：立即停止 try 區塊，**跳到 except** 執行

2. **執行 `except` 區塊**（只有出錯時才會執行）
   - 顯示友善訊息
   - 記錄錯誤
   - 使用預設值
   - 等等...

3. **繼續執行後面的程式碼**
   - 無論是否出錯，程式都會繼續往下執行（不會中斷）

### ✅ 重點
- `try` 放可能出錯的程式碼
- `except` 放錯誤處理邏輯
- 程式不會因為錯誤而中斷

## 📚 基本範例：安全的加法運算

以下範例模擬使用者輸入兩個數字並相加。

**情境**：使用者可能輸入中文數字「五」「十」，導致轉換失敗。

**解決方案**：使用 `try/except` 來處理轉換錯誤。

````python
# 範例 1：正確的輸入（阿拉伯數字）

# 模擬使用者輸入（在實際應用中，這會是 input() 的結果）
num1 = "5"
num2 = "10"

print("=== 情況一：正確輸入 ===")
try:
    result = int(num1) + int(num2)
    print(f"✅ 計算結果：{num1} + {num2} = {result}")
except:
    print("❌ 請輸入阿拉伯數字")

print("✅ 程式繼續執行")
````

```
=== 情況一：正確輸入 ===
✅ 計算結果：5 + 10 = 15

```

#### 情況二：錯誤輸入（num1="五", num2="十"）
```
=== 情況二：錯誤輸入 ===
❌ 請輸入阿拉伯數字
✅ 程式繼續執行
```

**重點**：
- 正確輸入：執行 try 區塊，顯示計算結果
- 錯誤輸入：跳到 except 區塊，顯示友善提示
- **無論哪種情況**：程式都繼續執行（顯示「程式繼續執行」）

````python
# 範例 2：錯誤的輸入（中文數字）

# 模擬使用者輸入錯誤資料
num1 = "五"
num2 = "十"

print("=== 情況二：錯誤輸入 ===")
try:
    result = int(num1) + int(num2)
    print(f"✅ 計算結果：{num1} + {num2} = {result}")
except:
    print("❌ 請輸入阿拉伯數字")

print("✅ 程式繼續執行")
````

## 指定例外類型（更精準的錯誤處理）

### ⚠️ 為什麼不要用「裸 except」

**裸 except**：不指定例外類型的 except
```python
try:
    # 程式碼
except:  # 這就是裸 except
    # 會接住所有錯誤
```

**問題**：
- 會接住**所有錯誤**，包括你不想接的（例如：KeyboardInterrupt）
- 無法針對不同錯誤給予不同處理
- 難以除錯（不知道到底發生什麼錯誤）

### ✅ 正確做法：指定例外類型

```python
try:
    # 程式碼
except ValueError:
    # 只處理 ValueError
except ZeroDivisionError:
    # 只處理 ZeroDivisionError
except:
    # 最後才用裸 except 保底（optional）
```

### 📝 常見例外類型
- **ValueError**：值錯誤（int("abc")）
- **ZeroDivisionError**：除以零（10 / 0）
- **NameError**：變數不存在（print(x) 但 x 沒定義）
- **TypeError**：類型錯誤（len(123)）
- **KeyError**：dict 找不到 key
- **IndexError**：list 索引超出範圍

### 🔄 比對順序
**Python 會由上往下比對，第一個符合的就執行**
```python
try:
    # 程式碼
except ZeroDivisionError:  # 先比對這個
    print("除數不可為零")
except ValueError:  # 再比對這個
    print("請輸入阿拉伯數字")
except:  # 最後才是這個（保底）
    print("發生未知錯誤")
```

## 📚 範例：多個 except（除法運算）

以下範例會處理兩種可能的錯誤：
1. **ValueError**：輸入不是數字
2. **ZeroDivisionError**：除數為零

先示範 **ZeroDivisionError** 的情況。

````python
# 範例 3：除法運算 - ZeroDivisionError

num1 = "15"
num2 = "0"  # 除數為零

print("=== 除法運算（除數為零）===")
try:
    result = int(num1) / int(num2)
    print(f"✅ 計算結果：{num1} / {num2} = {result}")
except ValueError:
    print("❌ 請輸入阿拉伯數字")
except ZeroDivisionError:
    print("❌ 除數不可為零")

print("✅ 程式繼續執行")
````

## 📚 範例：多個 except（ValueError）

現在示範 **ValueError** 的情況。

````python
# 範例 4：除法運算 - ValueError

num1 = "十五"  # 中文數字
num2 = "3"

print("=== 除法運算（輸入錯誤）===")
try:
    result = int(num1) / int(num2)
    print(f"✅ 計算結果：{num1} / {num2} = {result}")
except ValueError:
    print("❌ 請輸入阿拉伯數字")
except ZeroDivisionError:
    print("❌ 除數不可為零")

print("✅ 程式繼續執行")
````

## 常見例外類型總整理

以下是 Python 最常見的例外類型，**初學者必須掌握**！

| 例外類型 | 說明 | 範例 | 何時發生 |
|---------|------|------|---------|
| **SyntaxError** | 語法錯誤 | `if x = 5:` | 程式碼寫錯（執行前就會炸） |
| **IndentationError** | 縮排錯誤 | 忘記縮排或縮排不一致 | 程式碼格式錯誤（執行前就會炸） |
| **NameError** | 變數名稱錯誤 | `print(abc)`（abc 未定義） | 使用未定義的變數 |
| **TypeError** | 類型錯誤 | `len(123)` | 對錯誤的類型執行操作 |
| **ValueError** | 值錯誤 | `int("abc")` | 值不符合預期格式 |
| **ZeroDivisionError** | 除以零 | `10 / 0` | 數學運算錯誤 |
| **IndexError** | 索引錯誤 | `[1,2,3][99]` | list 索引超出範圍 |
| **KeyError** | Key 錯誤 | `d["key"]`（key 不存在） | dict 找不到指定 key |

### ⚠️ 特別注意

**SyntaxError 和 IndentationError**：
- 這兩個錯誤發生在**執行之前**（Python 讀取程式碼時）
- `try/except` **通常接不到**這兩個錯誤
- 解決方法：**寫程式時就要檢查語法**，不能靠 try/except

**其他錯誤**：
- 發生在**執行期間**（Runtime）
- `try/except` **可以接住**

### 💡 記憶技巧
- **NameError**：名字錯誤 → 變數不存在
- **TypeError**：類型錯誤 → 對錯誤的類型操作
- **ValueError**：值錯誤 → 值的格式不對
- **ZeroDivisionError**：除以零錯誤 → 字面意思
- **IndexError**：索引錯誤 → list 範圍外
- **KeyError**：Key 錯誤 → dict 找不到 key

## 📚 常見錯誤範例合集

以下示範如何接住各種常見錯誤，**每個錯誤都會被 try/except 安全接住**。

````python
# 常見錯誤範例合集

print("=== 常見錯誤示範 ===\n")

# 1. NameError：變數不存在
print("1. NameError（變數不存在）")
try:
    print(not_defined_var)
except NameError:
    print("   ❌ NameError：變數不存在")
    print("   💡 提示：請先定義變數再使用\n")

# 2. TypeError：類型錯誤
print("2. TypeError（類型錯誤）")
try:
    length = len(123)  # 數字沒有 len()
except TypeError:
    print("   ❌ TypeError：無法對數字使用 len()")
    print("   💡 提示：len() 只能用於字串、list、dict 等\n")

# 3. ValueError：值錯誤
print("3. ValueError（值錯誤）")
try:
    number = int("abc")
except ValueError:
    print("   ❌ ValueError：無法將 'abc' 轉換成整數")
    print("   💡 提示：請確認輸入的是數字\n")

# 4. IndexError：索引超出範圍
print("4. IndexError（索引超出範圍）")
try:
    my_list = [1, 2, 3]
    value = my_list[99]
except IndexError:
    print("   ❌ IndexError：索引 99 超出 list 範圍（只有 0-2）")
    print("   💡 提示：list 只有 3 個元素（index 0-2）\n")

# 5. ZeroDivisionError：除以零
print("5. ZeroDivisionError（除以零）")
try:
    result = 1 / 0
except ZeroDivisionError:
    print("   ❌ ZeroDivisionError：除數不可為零")
    print("   💡 提示：請檢查除數是否為 0\n")

print("=== 所有錯誤都被安全接住，程式正常結束 ===")
````

## 本章重點整理

### ✅ 核心觀念

1. **什麼是例外**
   - 程式執行期間發生的錯誤
   - 沒處理會導致程式中斷

2. **try / except 語法**
   ```python
   try:
       # 可能會出錯的程式碼
   except:
       # 處理錯誤
   ```

3. **指定例外類型（重要！）**
   - 不要用裸 except（太籠統）
   - 針對不同錯誤給予不同處理
   - 範例：`except ValueError`, `except ZeroDivisionError`

4. **except 比對順序**
   - Python 由上往下比對
   - 第一個符合的就執行
   - 建議把特定的錯誤放前面，裸 except 放最後（如果要用的話）

5. **不要濫用裸 except**
   - 會接住所有錯誤（包括你不想接的）
   - 難以除錯
   - 如果要保底，放在最後一個 except

### 📋 常見例外類型速查

| 例外 | 原因 | 範例 |
|-----|------|------|
| ValueError | 值格式錯誤 | `int("abc")` |
| ZeroDivisionError | 除以零 | `10 / 0` |
| NameError | 變數不存在 | `print(x)` 但 x 未定義 |
| TypeError | 類型錯誤 | `len(123)` |
| KeyError | dict key 不存在 | `d["key"]` 但 key 不在 d 中 |
| IndexError | list 索引超出範圍 | `[1,2,3][99]` |

### 🎯 實務建議

1. **使用者輸入一定要加 try/except**
2. **檔案操作、網路請求都要處理例外**
3. **給使用者友善的錯誤訊息**
4. **記錄錯誤日誌（進階）**
5. **不要用 try/except 隱藏程式錯誤**
