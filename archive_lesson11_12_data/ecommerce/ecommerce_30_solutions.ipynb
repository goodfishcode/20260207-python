## 題目 1 解答

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## 題目 2 解答

```python
df = pd.read_csv('ecommerce_sales.csv')
df['order_date'] = pd.to_datetime(df['order_date'])
```

## 題目 3 解答

```python
df.head(10)
```

## 題目 4 解答

```python
print(f"列數：{df.shape[0]}")
print(f"欄數：{df.shape[1]}")
```

## 題目 9 解答

```python
df.sort_values('price', ascending=False).head(5)
```

## 題目 10 解答

```python
electronics_count = len(df[df['category'] == 'Electronics'])
print(f"Electronics 訂單數：{electronics_count}")
```

## 題目 11 解答

```python
df['revenue'] = df['price'] * df['quantity']
df[['order_id', 'price', 'quantity', 'revenue']].head()
```

## 題目 12 解答

```python
df['profit'] = (df['price'] - df['cost']) * df['quantity']
df[['order_id', 'price', 'cost', 'quantity', 'profit']].head()
```

## 題目 13 解答

```python
df['profit_margin'] = df['profit'] / df['revenue']
df[['order_id', 'revenue', 'profit', 'profit_margin']].head()
```

## 題目 14 解答

```python
total_revenue = df['revenue'].sum()
total_profit = df['profit'].sum()
print(f"總營收：${total_revenue:,.2f}")
print(f"總利潤：${total_profit:,.2f}")
```

## 題目 15 解答

```python
df[df['revenue'] == df['revenue'].max()]
```

## 題目 16 解答

```python
category_revenue = df.groupby('category')['revenue'].sum().sort_values(ascending=False)
category_revenue
```

## 題目 17 解答

```python
region_stats = df.groupby('region').agg({
    'order_id': 'count',
    'revenue': 'mean'
})
region_stats.columns = ['訂單數', '平均營收']
region_stats
```

## 題目 18 解答

```python
channel_stats = df.groupby('channel')[['revenue', 'profit']].sum()
channel_stats.columns = ['總營收', '總利潤']
channel_stats
```

## 題目 19 解答

```python
pivot = df.pivot_table(values='revenue', index='region', columns='category', aggfunc='sum')
pivot
```

## 題目 20 解答

```python
top3_products = df.groupby('product')['revenue'].sum().sort_values(ascending=False).head(3)
top3_products
```

## 題目 21 解答

```python
revenue_mean = np.mean(df['revenue'])
revenue_median = np.median(df['revenue'])
revenue_std = np.std(df['revenue'])

print(f"平均值：${revenue_mean:,.2f}")
print(f"中位數：${revenue_median:,.2f}")
print(f"標準差：${revenue_std:,.2f}")
```

## 題目 22 解答

```python
revenue_90th = np.percentile(df['revenue'], 90)
print(f"90 分位數：${revenue_90th:,.2f}")
```

## 題目 23 解答

```python
df_sorted = df.sort_values('revenue', ascending=False)
top_20_percent_count = int(len(df) * 0.2)
top_20_percent_revenue = df_sorted.head(top_20_percent_count)['revenue'].sum()
total_revenue = df['revenue'].sum()
contribution = (top_20_percent_revenue / total_revenue) * 100

print(f"前 20% 訂單數：{top_20_percent_count}")
print(f"前 20% 訂單營收：${top_20_percent_revenue:,.2f}")
print(f"貢獻百分比：{contribution:.2f}%")
```

## 題目 24 解答

```python
df['order_level'] = np.where(df['revenue'] >= 10000, 'High',
                     np.where(df['revenue'] >= 5000, 'Medium', 'Low'))
df[['order_id', 'revenue', 'order_level']].head(10)
```

## 題目 25 解答

```python
level_stats = df.groupby('order_level').agg({
    'order_id': 'count',
    'revenue': 'sum'
})
level_stats.columns = ['訂單數', '總營收']
level_stats
```

## 題目 26 解答

```python
plt.figure(figsize=(10, 6))
plt.hist(df['revenue'], bins=20, edgecolor='black', alpha=0.7)
plt.title('營收分布直方圖', fontsize=14)
plt.xlabel('營收', fontsize=12)
plt.ylabel('訂單數', fontsize=12)
plt.grid(axis='y', alpha=0.3)
plt.show()
```

## 題目 27 解答

```python
plt.figure(figsize=(10, 6))
category_revenue = df.groupby('category')['revenue'].sum().sort_values(ascending=False)
category_revenue.plot(kind='bar', color='steelblue', edgecolor='black')
plt.title('各類別總營收', fontsize=14)
plt.xlabel('類別', fontsize=12)
plt.ylabel('總營收', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

## 題目 28 解答

```python
daily_revenue = df.groupby('order_date')['revenue'].sum().reset_index()

plt.figure(figsize=(12, 6))
plt.plot(daily_revenue['order_date'], daily_revenue['revenue'], linewidth=2, color='green')
plt.title('每日營收趨勢', fontsize=14)
plt.xlabel('日期', fontsize=12)
plt.ylabel('營收', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## 題目 29 解答

```python
plt.figure(figsize=(8, 6))
channel_avg = df.groupby('channel')['revenue'].mean().sort_values(ascending=False)
channel_avg.plot(kind='bar', color='orange', edgecolor='black')
plt.title('各通路平均營收比較', fontsize=14)
plt.xlabel('通路', fontsize=12)
plt.ylabel('平均營收', fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

## 題目 30 解答

```python
# 1. 營收最高的類別
top_category = df.groupby('category')['revenue'].sum().idxmax()
top_category_revenue = df.groupby('category')['revenue'].sum().max()

# 2. 表現最好的地區
top_region = df.groupby('region')['revenue'].sum().idxmax()
top_region_revenue = df.groupby('region')['revenue'].sum().max()

# 3. 高價值訂單（前 10%）的平均營收
top_10_percent_count = int(len(df) * 0.1)
df_sorted = df.sort_values('revenue', ascending=False)
top_10_avg = df_sorted.head(top_10_percent_count)['revenue'].mean()

print("=== 綜合分析結果 ===")
print(f"1. 營收最高的類別：{top_category}（${top_category_revenue:,.2f}）")
print(f"2. 表現最好的地區：{top_region}（${top_region_revenue:,.2f}）")
print(f"3. 高價值訂單平均營收：${top_10_avg:,.2f}")
print()
print("=== 商業建議 ===")
print(f"建議加強 {top_category} 類別在 {top_region} 地區的行銷投資，")
print(f"並針對高價值客戶群建立 VIP 計畫以提升客戶忠誠度。")
```
