## ç¬¬ 7 é¡Œï¼šè¨ˆç®—ç¸½å’Œ

## ç¬¬ 17 é¡Œï¼šè¨ˆç®—éšä¹˜

**é¡Œç›®ï¼š** æ’°å¯«ä¸€å€‹åç‚º `factorial(n)` çš„å‡½å¼ï¼Œæ¥æ”¶ä¸€å€‹éè² æ•´æ•¸ `n`ï¼Œ**å›å‚³** n çš„éšä¹˜ï¼ˆn!ï¼‰ã€‚

````python
# è§£ç­”
def factorial(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

# æ¸¬è©¦
print(factorial(5))    # 120 (5Ã—4Ã—3Ã—2Ã—1)
print(factorial(3))    # 6 (3Ã—2Ã—1)
print(factorial(0))    # 1

# è¼¸å‡ºï¼š
# 120
# 6
# 1
````
