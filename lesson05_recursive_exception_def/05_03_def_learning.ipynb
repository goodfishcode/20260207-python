````python
# ä½¿ç”¨ return ç‰ˆæœ¬
def check_fever(t):
    """
    æª¢æŸ¥é«”æº«æ˜¯å¦ç™¼ç‡’
    åƒæ•¸ t: é«”æº«ï¼ˆæ”æ°ï¼‰
    å›å‚³: Trueï¼ˆç™¼ç‡’ï¼‰æˆ– Falseï¼ˆæ­£å¸¸ï¼‰
    """
    if t >= 37.5:
        return True  # ç™¼ç‡’
    else:
        return False  # æ­£å¸¸

# æ¸¬è©¦ä¸¦ä½¿ç”¨å›å‚³å€¼
temp1 = 36.5
result1 = check_fever(temp1)
print(f"é«”æº« {temp1}Â°Cï¼Œç™¼ç‡’ï¼Ÿ{result1}")

temp2 = 38.2
result2 = check_fever(temp2)
print(f"é«”æº« {temp2}Â°Cï¼Œç™¼ç‡’ï¼Ÿ{result2}")

# å¯ä»¥æ ¹æ“šå›å‚³å€¼åšä¸åŒè™•ç†
print("\n=== æ ¹æ“šå›å‚³å€¼åšåˆ¤æ–· ===")
my_temp = 37.8
if check_fever(my_temp):
    print(f"é«”æº« {my_temp}Â°Cï¼Œéœ€è¦ä¼‘æ¯ï¼")
else:
    print(f"é«”æº« {my_temp}Â°Cï¼Œå¥åº·è‰¯å¥½ï¼")

# è¼¸å‡ºï¼š
# é«”æº« 36.5Â°Cï¼Œç™¼ç‡’ï¼ŸFalse
# é«”æº« 38.2Â°Cï¼Œç™¼ç‡’ï¼ŸTrue
# 
# === æ ¹æ“šå›å‚³å€¼åšåˆ¤æ–· ===
# é«”æº« 37.8Â°Cï¼Œéœ€è¦ä¼‘æ¯ï¼
````

### ç¯„ä¾‹

- 2020ï¼šèƒ½è¢« 4 æ•´é™¤ï¼Œä¸èƒ½è¢« 100 æ•´é™¤ â†’ âœ… é–å¹´
- 1900ï¼šèƒ½è¢« 4 æ•´é™¤ï¼Œèƒ½è¢« 100 æ•´é™¤ï¼Œä¸èƒ½è¢« 400 æ•´é™¤ â†’ âŒ å¹³å¹´
- 2000ï¼šèƒ½è¢« 400 æ•´é™¤ â†’ âœ… é–å¹´

### å¯¦ä½œå‡½å¼

````python
def check_leap_year(year):
    """
    åˆ¤æ–·æ˜¯å¦ç‚ºé–å¹´
    åƒæ•¸ year: å¹´ä»½ï¼ˆæ•´æ•¸ï¼‰
    å›å‚³: Trueï¼ˆé–å¹´ï¼‰æˆ– Falseï¼ˆå¹³å¹´ï¼‰
    """
    # è¦å‰‡ 1ï¼šèƒ½è¢« 400 æ•´é™¤ â†’ é–å¹´
    if year % 400 == 0:
        return True
    
    # è¦å‰‡ 2ï¼šèƒ½è¢« 100 æ•´é™¤ï¼ˆä½†ä¸èƒ½è¢« 400 æ•´é™¤ï¼‰â†’ å¹³å¹´
    if year % 100 == 0:
        return False
    
    # è¦å‰‡ 3ï¼šèƒ½è¢« 4 æ•´é™¤ â†’ é–å¹´
    if year % 4 == 0:
        return True
    
    # å…¶ä»–æƒ…æ³ â†’ å¹³å¹´
    return False

# æ¸¬è©¦å¤šå€‹å¹´ä»½
test_years = [2020, 2021, 1900, 2000, 2024, 2100]

print("=== é–å¹´åˆ¤æ–·çµæœ ===")
for year in test_years:
    is_leap = check_leap_year(year)
    result = "é–å¹´" if is_leap else "å¹³å¹´"
    print(f"{year} å¹´ï¼š{result}")

# è¼¸å‡ºï¼š
# === é–å¹´åˆ¤æ–·çµæœ ===
# 2020 å¹´ï¼šé–å¹´
# 2021 å¹´ï¼šå¹³å¹´
# 1900 å¹´ï¼šå¹³å¹´
# 2000 å¹´ï¼šé–å¹´
# 2024 å¹´ï¼šé–å¹´
# 2100 å¹´ï¼šå¹³å¹´
````

````python
def count_vowels(text):
    """
    è¨ˆç®—å­—ä¸²ä¸­çš„æ¯éŸ³æ•¸é‡
    æ¯éŸ³åŒ…æ‹¬ï¼ša, e, i, o, uï¼ˆä¸å€åˆ†å¤§å°å¯«ï¼‰
    """
    vowels = "aeiouAEIOU"
    count = 0
    
    for char in text:
        if char in vowels:
            count += 1
    
    return count

# æ¸¬è©¦
word1 = "Hello"
word2 = "Python"
word3 = "Programming"

print(f'"{word1}" çš„æ¯éŸ³æ•¸é‡ï¼š{count_vowels(word1)}')
print(f'"{word2}" çš„æ¯éŸ³æ•¸é‡ï¼š{count_vowels(word2)}')
print(f'"{word3}" çš„æ¯éŸ³æ•¸é‡ï¼š{count_vowels(word3)}')

# å•é¡Œï¼šé€™ä¸‰å€‹å–®å­—å„æœ‰å¹¾å€‹æ¯éŸ³ï¼Ÿ
# å…ˆè‡ªå·±æƒ³ä¸€æƒ³ï¼Œå†åŸ·è¡Œç¨‹å¼ç¢¼é©—è­‰ï¼
````
