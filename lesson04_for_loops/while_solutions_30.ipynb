## 題目 14（解答）

```python
# 密碼驗證（最多 3 次機會）
# 使用測試資料模擬
test_inputs = ["abc", "def", "python123"]
test_index = 0

correct_password = "python123"
attempts = 0
max_attempts = 3
success = False

while attempts < max_attempts:
    password = test_inputs[test_index]
    test_index += 1
    attempts += 1
    
    if password == correct_password:
        print("登入成功！")
        success = True
        break
    else:
        remaining = max_attempts - attempts
        if remaining > 0:
            print(f"密碼錯誤，剩餘 {remaining} 次機會")

if not success:
    print("帳號已鎖定")
```

```python
# 統計數字中每個位數（巢狀概念）
numbers = [5, 42, 108, 9876]

index = 0
while index < len(numbers):
    num = numbers[index]
    original = num
    count = 0
    
    # 內層迴圈：計算位數
    while num > 0:
        count += 1
        num //= 10
    
    print(f"{original} 有 {count} 位數")
    index += 1
```

## 題目 26（解答）

```python
# 質數判斷（容易陷入錯誤）
n = 17

is_prime = True
divisor = 2

while divisor < n:
    if n % divisor == 0:
        is_prime = False
        break  # 找到因數後結束
    divisor += 1  # 【重要】記得更新 divisor

if is_prime:
    print(f"{n} 是質數")
else:
    print(f"{n} 不是質數")
```

## 題目 28（解答）

```python
# 移除串列中的特定元素（容易陷入錯誤）
numbers = [1, 2, 3, 2, 4, 2, 5]
target = 2

index = 0
while index < len(numbers):
    if numbers[index] == target:
        numbers.pop(index)  # 移除元素
        # 【重要】移除後不增加索引，因為後面的元素會前移
    else:
        index += 1  # 沒有移除時才增加索引

print(numbers)
```

## 題目 29（解答）

```python
# 費氏數列生成器
n = 10

fibonacci = []

if n >= 1:
    fibonacci.append(1)
if n >= 2:
    fibonacci.append(1)

count = 2
while count < n:
    next_value = fibonacci[count - 1] + fibonacci[count - 2]
    fibonacci.append(next_value)
    count += 1

print(fibonacci)
```

## 題目 30（解答）

```python
# 文字冒險遊戲（綜合應用）
# 使用測試資料模擬
test_inputs = [2, 2, 1, 2, 3]
test_index = 0

hp = 100
score = 0

while True:
    # 顯示目前狀態
    print(f"\nHP: {hp} | 分數: {score}")
    print("1. 休息（+20 HP）")
    print("2. 探險（-30 HP，+10 分）")
    print("3. 結束遊戲")
    
    # 讀取選擇
    choice = test_inputs[test_index]
    test_index += 1
    
    if choice == 1:
        # 休息
        hp += 20
        if hp > 100:
            hp = 100
        print("休息恢復體力")
    elif choice == 2:
        # 探險
        hp -= 30
        score += 10
        print("探險成功！獲得 10 分")
    elif choice == 3:
        # 結束遊戲
        break
    
    # 檢查 HP
    if hp <= 0:
        print("\nHP 歸零，遊戲結束！")
        break

print(f"遊戲結束！總分: {score}")
```
