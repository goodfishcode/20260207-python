## 06_Accumulatorsï¼šç´¯åŠ å™¨æ¨¡å¼

å­¸ç¿’å¦‚ä½•åœ¨è¿´åœˆä¸­ç´¯ç©çµæœï¼ˆsum, count, max, min, avg, running totalï¼‰

### ç·´ç¿’ 6.1ï¼šè¨ˆç®—ç¸½å’Œèˆ‡å¹³å‡
è¨ˆç®—ä¸²åˆ—çš„ç¸½å’Œå’Œå¹³å‡å€¼

````python
scores = [85, 92, 78, 90, 88]
total = 0
for score in scores:
    total += score
average = total / len(scores)
print(f"ç¸½å’Œ: {total}")
print(f"å¹³å‡: {average}")
````

### ç·´ç¿’ 6.2ï¼šæ‰¾å‡ºæœ€å¤§å€¼å’Œæœ€å°å€¼
ä¸ä½¿ç”¨å…§å»ºå‡½æ•¸ï¼Œæ‰¾å‡ºä¸²åˆ—ä¸­çš„æœ€å¤§å€¼å’Œæœ€å°å€¼

````python
numbers = [23, 45, 12, 67, 34, 89, 56]
max_val = numbers[0]
min_val = numbers[0]

for num in numbers:
    if num > max_val:
        max_val = num
    if num < min_val:
        min_val = num

print(f"æœ€å¤§å€¼: {max_val}")
print(f"æœ€å°å€¼: {min_val}")
````

### ç·´ç¿’ 6.3ï¼šè¨ˆæ•¸å™¨
è¨ˆç®—å­—ä¸²ä¸­æœ‰å¤šå°‘å€‹å¤§å¯«å­—æ¯å’Œå°å¯«å­—æ¯

````python
text = "Hello World Python Programming"
uppercase_count = 0
lowercase_count = 0

for char in text:
    if char.isupper():
        uppercase_count += 1
    elif char.islower():
        lowercase_count += 1

print(f"å¤§å¯«å­—æ¯æ•¸é‡: {uppercase_count}")
print(f"å°å¯«å­—æ¯æ•¸é‡: {lowercase_count}")
````

### ç·´ç¿’ 6.4ï¼šRunning Total

### ç·´ç¿’ 7.3ï¼šçŸ©å½¢æ¡†æ¡†

### å°ˆé¡Œ 10.1ï¼šå¯†ç¢¼å¼·åº¦æª¢æŸ¥å™¨
æª¢æŸ¥å¯†ç¢¼æ˜¯å¦ç¬¦åˆä»¥ä¸‹æ¢ä»¶ï¼š
- é•·åº¦è‡³å°‘ 8 å€‹å­—å…ƒ
- åŒ…å«è‡³å°‘ä¸€å€‹æ•¸å­—
- åŒ…å«è‡³å°‘ä¸€å€‹å¤§å¯«å­—æ¯
- åŒ…å«è‡³å°‘ä¸€å€‹å°å¯«å­—æ¯

````python
password = "MyPass123"

# åˆå§‹åŒ–æª¢æŸ¥æ¨™è¨˜
has_digit = False
has_upper = False
has_lower = False

# æª¢æŸ¥é•·åº¦
if len(password) >= 8:
    print("âœ“ é•·åº¦è¶³å¤ ï¼ˆè‡³å°‘ 8 å€‹å­—å…ƒï¼‰")
else:
    print("âœ— é•·åº¦ä¸è¶³")

# æª¢æŸ¥æ¯å€‹å­—å…ƒ
for char in password:
    if char.isdigit():
        has_digit = True
    elif char.isupper():
        has_upper = True
    elif char.islower():
        has_lower = True

# å°å‡ºçµæœ
if has_digit:
    print("âœ“ åŒ…å«æ•¸å­—")
else:
    print("âœ— ç¼ºå°‘æ•¸å­—")

if has_upper:
    print("âœ“ åŒ…å«å¤§å¯«å­—æ¯")
else:
    print("âœ— ç¼ºå°‘å¤§å¯«å­—æ¯")

if has_lower:
    print("âœ“ åŒ…å«å°å¯«å­—æ¯")
else:
    print("âœ— ç¼ºå°‘å°å¯«å­—æ¯")

# ç¸½çµ
if len(password) >= 8 and has_digit and has_upper and has_lower:
    print("\nå¯†ç¢¼å¼·åº¦ï¼šå¼· ğŸ’ª")
else:
    print("\nå¯†ç¢¼å¼·åº¦ï¼šå¼± âš ï¸")
````

### å°ˆé¡Œ 10.2ï¼šè³¼ç‰©è»Šç¸½åƒ¹è¨ˆç®—

````python
cart = [
    {'name': 'Apple', 'price': 30, 'quantity': 5},
    {'name': 'Banana', 'price': 20, 'quantity': 3},
    {'name': 'Orange', 'price': 25, 'quantity': 4}
]

total = 0
print("è³¼ç‰©è»Šæ˜ç´°ï¼š")
print("-" * 40)

for item in cart:
    item_total = item['price'] * item['quantity']
    total += item_total
    print(f"{item['name']:10} ${item['price']:3} x {item['quantity']} = ${item_total}")

print("-" * 40)
print(f"å°è¨ˆ: ${total}")

# å¥—ç”¨æŠ˜æ‰£
if total >= 1000:
    discount = total * 0.1
    final_total = total * 0.9
    print(f"æŠ˜æ‰£ (10%): -${discount:.2f}")
    print(f"ç¸½è¨ˆ: ${final_total:.2f}")
else:
    print(f"ç¸½è¨ˆ: ${total}")
    print(f"(å†æ¶ˆè²» ${1000 - total} å…ƒå³å¯äº« 9 æŠ˜å„ªæƒ )")
````

### å°ˆé¡Œ 10.3ï¼šè³ªæ•¸åˆ¤æ–·
åˆ¤æ–·ä¸€å€‹æ•¸å­—æ˜¯å¦ç‚ºè³ªæ•¸

````python
number = 17
is_prime = True

if number <= 1:
    is_prime = False
else:
    # æª¢æŸ¥å¾ 2 åˆ° number-1 æ˜¯å¦æœ‰å› æ•¸
    for i in range(2, number):
        if number % i == 0:
            is_prime = False
            print(f"{number} å¯ä»¥è¢« {i} æ•´é™¤")
            break

if is_prime:
    print(f"{number} æ˜¯è³ªæ•¸ âœ“")
else:
    print(f"{number} ä¸æ˜¯è³ªæ•¸ âœ—")
````

### å°ˆé¡Œ 10.4ï¼šæ–‡å­—çµ±è¨ˆåˆ†æ

- å­éŸ³æ•¸é‡

````python
text = "Hello World 123 Python Programming 456"

vowel_count = 0
consonant_count = 0
digit_count = 0
vowels = "aeiouAEIOU"

# è¨ˆç®—å­—æ•¸ï¼ˆä»¥ç©ºæ ¼åˆ†éš”ï¼‰
words = text.split()
word_count = len(words)

# åˆ†ææ¯å€‹å­—å…ƒ
for char in text:
    if char.isalpha():
        if char in vowels:
            vowel_count += 1
        else:
            consonant_count += 1
    elif char.isdigit():
        digit_count += 1

print("æ–‡å­—çµ±è¨ˆåˆ†æ")
print("=" * 30)
print(f"æ–‡å­—å…§å®¹: {text}")
print(f"å­—æ•¸: {word_count}")
print(f"æ¯éŸ³æ•¸é‡: {vowel_count}")
print(f"å­éŸ³æ•¸é‡: {consonant_count}")
print(f"æ•¸å­—æ•¸é‡: {digit_count}")
print(f"å­—æ¯ç¸½æ•¸: {vowel_count + consonant_count}")
````
