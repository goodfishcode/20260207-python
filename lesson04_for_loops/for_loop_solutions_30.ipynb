## ç¬¬å…«ç« ï¼šé€²éšæ‡‰ç”¨ (Q29-Q30)

### Q29: è³ªæ•¸åˆ¤æ–·

**æç¤ºï¼š** å¾ 2 åˆ° n-1 æª¢æŸ¥æ˜¯å¦æœ‰æ•¸å­—èƒ½æ•´é™¤ nã€‚è‹¥æœ‰å‰‡ä¸æ˜¯è³ªæ•¸ï¼Œå¯ä»¥ä½¿ç”¨ break ææ—©çµæŸã€‚

### Q30: è²»æ°æ•¸åˆ—ç”Ÿæˆå™¨

**é¡Œç›®æè¿°ï¼š**  
ç”Ÿæˆå‰ n é …è²»æ°æ•¸åˆ—ï¼ˆFibonacci sequenceï¼‰ã€‚è²»æ°æ•¸åˆ—å®šç¾©ç‚ºï¼šç¬¬ 1 é …å’Œç¬¬ 2 é …éƒ½æ˜¯ 1ï¼Œä¹‹å¾Œæ¯é …éƒ½æ˜¯å‰å…©é …ä¹‹å’Œã€‚
