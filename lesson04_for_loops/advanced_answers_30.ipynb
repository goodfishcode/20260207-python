### Q20ï¼šåˆ¤æ–·é˜¿å§†æ–¯å£¯æ•¸ï¼ˆArmstrong Numberï¼‰

**é¡Œç›®ï¼š** é˜¿å§†æ–¯å£¯æ•¸æ˜¯æŒ‡ä¸€å€‹ n ä½æ•¸ï¼Œå…¶å„ä½æ•¸å­—çš„ n æ¬¡æ–¹å’Œç­‰æ–¼è©²æ•¸æœ¬èº«ã€‚

```python
# Q19 è§£ç­”ï¼ˆè¼¾è½‰ç›¸é™¤æ³•ï¼‰
a = 48
b = 18

# ä¿å­˜åŸå§‹å€¼ç”¨æ–¼è¼¸å‡º
original_a, original_b = a, b

while b != 0:
    temp = b
    b = a % b
    a = temp

gcd = a
print(f"{original_a} å’Œ {original_b} çš„æœ€å¤§å…¬å› æ•¸æ˜¯: {gcd}")
```

**é¡Œç›®ï¼š** ç”Ÿæˆå¸•æ–¯å¡ä¸‰è§’å½¢ï¼ˆPascal's Triangleï¼‰çš„å‰ n åˆ—ã€‚

### Q27ï¼šæ¨¡æ“¬ç°¡æ˜“ ATM ç³»çµ±

**é¡Œç›®ï¼š** å»ºç«‹ä¸€å€‹ç°¡å–®çš„ ATM ææ¬¾æ©Ÿæ¨¡æ“¬ç³»çµ±ï¼Œæ”¯æ´å­˜æ¬¾ã€ææ¬¾ã€æŸ¥è©¢é¤˜é¡ã€‚

```python
# Q26 è§£ç­”
n = 5
triangle = []

for i in range(n):
    row = [1]  # æ¯åˆ—çš„ç¬¬ä¸€å€‹å…ƒç´ éƒ½æ˜¯ 1
    
    if i > 0:
        # è¨ˆç®—ä¸­é–“çš„å…ƒç´ 
        for j in range(1, i):
            row.append(triangle[i-1][j-1] + triangle[i-1][j])
        row.append(1)  # æ¯åˆ—çš„æœ€å¾Œä¸€å€‹å…ƒç´ éƒ½æ˜¯ 1
    
    triangle.append(row)
    print(row)
```

### Q29ï¼šæ‰¾å‡ºæœ€é•·é€£çºŒéå¢å­åºåˆ—
