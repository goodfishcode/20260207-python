## 1ï¸âƒ£ Class èˆ‡ Object æ˜¯ä»€éº¼ï¼Ÿ

### ğŸ—ï¸ ç”¨ã€Œå’–å•¡åº—ã€ä¾†ç†è§£

æƒ³åƒä½ èµ°é€²ä¸€å®¶å’–å•¡åº—ï¼Œé»äº†ä¸€æ¯æ‹¿éµï¼š

- **Classï¼ˆé¡åˆ¥ï¼‰** = å’–å•¡çš„ã€Œé£Ÿè­œã€æˆ–ã€Œè¨­è¨ˆåœ–ã€
  - å®šç¾©äº†å’–å•¡æœ‰ä»€éº¼ç‰¹æ€§ï¼ˆsizeã€priceã€nameï¼‰

- **Objectï¼ˆç‰©ä»¶ï¼‰** = å¯¦éš›åšå‡ºä¾†çš„ã€Œé‚£ä¸€æ¯å’–å•¡ã€
  - ä½ çš„æ‹¿éµæ˜¯ä¸€å€‹ object

**é‡é»**ï¼š
- ä¸€å€‹ Class å¯ä»¥ç”¢ç”Ÿç„¡æ•¸å€‹ Object
- æ¯å€‹ Object æœ‰è‡ªå·±çš„è³‡æ–™ï¼ˆä½ çš„æ‹¿éµæ˜¯å¤§æ¯ï¼Œä»–çš„ç¾å¼æ˜¯ä¸­æ¯ï¼‰

## 2ï¸âƒ£ `__init__` èˆ‡ `self` æ˜¯ä»€éº¼ï¼Ÿ

### ğŸ¼ ç”¨ã€Œå‡ºç”Ÿç™»è¨˜ã€ä¾†ç†è§£

ç•¶ä¸€å€‹å¬°å…’å‡ºç”Ÿæ™‚ï¼Œæœƒç™»è¨˜ï¼šå§“åã€æ€§åˆ¥ã€ç”Ÿæ—¥ç­‰ã€Œä¸€å‡ºç”Ÿå°±æœ‰çš„è³‡æ–™ã€ã€‚

- **`__init__`** = ç‰©ä»¶çš„ã€Œå‡ºç”Ÿç™»è¨˜ç¨‹åºã€
  - æ¯æ¬¡å»ºç«‹æ–° object æ™‚æœƒè‡ªå‹•åŸ·è¡Œ
  - è² è²¬è¨­å®šç‰©ä»¶çš„åˆå§‹è³‡æ–™

- **`self`** = ã€Œé€™å€‹ç‰©ä»¶è‡ªå·±ã€
  - `self.name` è¡¨ç¤ºã€Œé€™å€‹ç‰©ä»¶çš„ nameã€
  - è®“æ¯å€‹ object éƒ½æœ‰è‡ªå·±ç¨ç«‹çš„è³‡æ–™

**ç™½è©±æ–‡**ï¼š
- `__init__` å°±åƒå¬°å…’çš„å‡ºç”Ÿè­‰æ˜å¡«å¯«æµç¨‹
- `self` å°±æ˜¯åœ¨èªªã€Œæˆ‘è‡ªå·±çš„è³‡æ–™ã€

````python
class Person:
    def __init__(self, name, age):
        # __init__ æœƒåœ¨å»ºç«‹ç‰©ä»¶æ™‚è‡ªå‹•åŸ·è¡Œ
        print(f"ğŸ‰ æ­£åœ¨å»ºç«‹ {name} çš„è³‡æ–™...")
        self.name = name  # self.name è¡¨ç¤ºã€Œé€™å€‹äººçš„åå­—ã€
        self.age = age    # self.age è¡¨ç¤ºã€Œé€™å€‹äººçš„å¹´é½¡ã€
    
    def introduce(self):
        # åœ¨æ–¹æ³•è£¡ä¹Ÿç”¨ self ä¾†å­˜å–è‡ªå·±çš„è³‡æ–™
        print(f"å—¨ï¼æˆ‘æ˜¯ {self.name}ï¼Œä»Šå¹´ {self.age} æ­²ã€‚")

# å»ºç«‹å…©å€‹äºº
person1 = Person("å°æ˜", 25)  # é€™è£¡æœƒè‡ªå‹•å‘¼å« __init__
person2 = Person("å°è¯", 30)  # åˆæœƒå†å‘¼å«ä¸€æ¬¡ __init__

print("\n--- è‡ªæˆ‘ä»‹ç´¹ ---")
person1.introduce()
person2.introduce()

print("\nâœ¨ æ³¨æ„ï¼šæ¯å€‹äººéƒ½æœ‰è‡ªå·±ç¨ç«‹çš„ name å’Œ ageï¼")
````

## 3ï¸âƒ£ å±¬æ€§ï¼ˆAttributeï¼‰vs æ–¹æ³•ï¼ˆMethodï¼‰

### ğŸ“± ç”¨ã€Œæ‰‹æ©Ÿã€ä¾†ç†è§£

ä¸€æ”¯æ‰‹æ©Ÿæœ‰ï¼š

- **å±¬æ€§ï¼ˆAttributeï¼‰** = æ‰‹æ©Ÿçš„ã€Œç‹€æ…‹ã€æˆ–ã€Œç‰¹å¾µã€
  - ä¾‹å¦‚ï¼šé›»é‡ã€éŸ³é‡ã€å“ç‰Œã€é¡è‰²
  - é€™äº›æ˜¯ã€Œåè©ã€ï¼Œæè¿°æ‰‹æ©Ÿã€Œæ˜¯ä»€éº¼æ¨£å­ã€

- **æ–¹æ³•ï¼ˆMethodï¼‰** = æ‰‹æ©Ÿçš„ã€ŒåŠŸèƒ½ã€æˆ–ã€Œè¡Œç‚ºã€
  - ä¾‹å¦‚ï¼šèª¿æ•´éŸ³é‡ã€æ‰“é›»è©±ã€æ‹ç…§
  - é€™äº›æ˜¯ã€Œå‹•è©ã€ï¼Œæè¿°æ‰‹æ©Ÿã€Œèƒ½åšä»€éº¼ã€

**è¨˜æ†¶æ³•**ï¼š
- å±¬æ€§ = åè©ï¼ˆéŸ³é‡**æ˜¯**å¤šå°‘ï¼Ÿï¼‰
- æ–¹æ³• = å‹•è©ï¼ˆ**èª¿æ•´**éŸ³é‡ã€**æ‰“**é›»è©±ï¼‰

````python
class Phone:
    def __init__(self, brand, color):
        # å±¬æ€§ï¼ˆAttributesï¼‰â€” æ‰‹æ©Ÿçš„ç‰¹å¾µ
        self.brand = brand
        self.color = color
        self.volume = 50     # é è¨­éŸ³é‡ 50
        self.battery = 100   # é è¨­é›»é‡ 100%
    
    # æ–¹æ³•ï¼ˆMethodsï¼‰â€” æ‰‹æ©Ÿçš„åŠŸèƒ½
    def increase_volume(self):
        if self.volume < 100:
            self.volume += 10
            print(f"ğŸ”Š éŸ³é‡å¢åŠ åˆ° {self.volume}")
        else:
            print("âš ï¸ éŸ³é‡å·²é”æœ€å¤§å€¼ï¼")
    
    def decrease_volume(self):
        if self.volume > 0:
            self.volume -= 10
            print(f"ğŸ”‰ éŸ³é‡é™ä½åˆ° {self.volume}")
        else:
            print("âš ï¸ å·²éœéŸ³ï¼")
    
    def show_status(self):
        print(f"ğŸ“± {self.brand} ({self.color})")
        print(f"   ğŸ”‹ é›»é‡ï¼š{self.battery}%")
        print(f"   ğŸ”Š éŸ³é‡ï¼š{self.volume}")

# å»ºç«‹ä¸€æ”¯æ‰‹æ©Ÿ
my_phone = Phone("iPhone", "é»‘è‰²")

# ä½¿ç”¨å±¬æ€§å’Œæ–¹æ³•
my_phone.show_status()
print("\n--- æ¸¬è©¦éŸ³é‡èª¿æ•´ ---")
my_phone.increase_volume()
my_phone.increase_volume()
my_phone.decrease_volume()
````

## 4ï¸âƒ£ å°è£ï¼ˆEncapsulationï¼‰

### ğŸ¦ ç”¨ã€ŒéŠ€è¡Œå¸³æˆ¶ã€ä¾†ç†è§£

æƒ³åƒä½ å»éŠ€è¡Œï¼š

- âŒ ä½ **ä¸èƒ½**ç›´æ¥ä¿®æ”¹è³‡æ–™åº«è£¡çš„é¤˜é¡æ•¸å­—
- âœ… ä½ **åªèƒ½**é€éã€Œå­˜æ¬¾ã€æˆ–ã€Œææ¬¾ã€ä¾†æ”¹è®Šé¤˜é¡

é€™å°±æ˜¯**å°è£**çš„æ¦‚å¿µï¼š

1. **éš±è—å…§éƒ¨è³‡æ–™**ï¼ˆç”¨ `_` é–‹é ­è¡¨ç¤ºã€Œè«‹ä¸è¦ç›´æ¥æ”¹æˆ‘ã€ï¼‰
2. **æä¾›å®‰å…¨çš„æ–¹æ³•**ä¾†å­˜å–è³‡æ–™ï¼ˆdepositã€withdrawï¼‰
3. **åŠ å…¥é©—è­‰é‚è¼¯**ï¼ˆä¸èƒ½ææ¬¾è¶…éé¤˜é¡ï¼‰

**ç‚ºä»€éº¼è¦å°è£ï¼Ÿ**
- é˜²æ­¢è³‡æ–™è¢«äº‚æ”¹ï¼ˆä¸èƒ½ç›´æ¥æ”¹é¤˜é¡ç‚º 100 è¬ï¼‰
- åŠ å…¥å•†æ¥­é‚è¼¯ï¼ˆææ¬¾è¦æª¢æŸ¥é¤˜é¡å¤ ä¸å¤ ï¼‰
- ä¿è­·è³‡æ–™å®‰å…¨

````python
class BankAccount:
    def __init__(self, owner, initial_balance=0):
        self.owner = owner
        # _balance ç”¨åº•ç·šé–‹é ­ï¼Œè¡¨ç¤ºã€Œé€™æ˜¯å…§éƒ¨è³‡æ–™ï¼Œè«‹ä¸è¦ç›´æ¥æ”¹ã€
        self._balance = initial_balance
    
    # å­˜æ¬¾æ–¹æ³•ï¼ˆå”¯ä¸€å®‰å…¨æ”¹è®Šé¤˜é¡çš„æ–¹å¼ï¼‰
    def deposit(self, amount):
        if amount > 0:
            self._balance += amount
            print(f"ğŸ’° å­˜æ¬¾æˆåŠŸï¼å­˜å…¥ ${amount}ï¼Œç›®å‰é¤˜é¡ï¼š${self._balance}")
        else:
            print("âš ï¸ å­˜æ¬¾é‡‘é¡å¿…é ˆå¤§æ–¼ 0")
    
    # ææ¬¾æ–¹æ³•ï¼ˆæœ‰é©—è­‰é‚è¼¯ï¼‰
    def withdraw(self, amount):
        if amount <= 0:
            print("âš ï¸ ææ¬¾é‡‘é¡å¿…é ˆå¤§æ–¼ 0")
        elif amount > self._balance:
            print(f"âŒ é¤˜é¡ä¸è¶³ï¼ç›®å‰é¤˜é¡ï¼š${self._balance}")
        else:
            self._balance -= amount
            print(f"ğŸ’¸ ææ¬¾æˆåŠŸï¼æå‡º ${amount}ï¼Œå‰©é¤˜é¤˜é¡ï¼š${self._balance}")
    
    # å”¯è®€å±¬æ€§ï¼ˆåªèƒ½çœ‹ï¼Œä¸èƒ½æ”¹ï¼‰
    @property
    def balance(self):
        return self._balance

# å»ºç«‹å¸³æˆ¶
account = BankAccount("å°æ˜", 1000)

print(f"å¸³æˆ¶æ“æœ‰è€…ï¼š{account.owner}")
print(f"ç›®å‰é¤˜é¡ï¼š${account.balance}")  # ç”¨ @property å¯ä»¥åƒå±¬æ€§ä¸€æ¨£å­˜å–

print("\n--- æ¸¬è©¦äº¤æ˜“ ---")
account.deposit(500)
account.withdraw(300)
account.withdraw(2000)  # æœƒå¤±æ•—ï¼Œé¤˜é¡ä¸è¶³

print(f"\nğŸ’¡ æ³¨æ„ï¼šä¸èƒ½ç›´æ¥å¯« account.balance = 9999")
print(f"   å¿…é ˆé€é deposit/withdraw ä¾†æ”¹è®Šé¤˜é¡ï¼")
````

## 5ï¸âƒ£ ç¹¼æ‰¿ï¼ˆInheritanceï¼‰

### ğŸ¾ ç”¨ã€Œå‹•ç‰©ã€ä¾†ç†è§£

æƒ³åƒå‹•ç‰©ä¸–ç•Œï¼š

- **æ‰€æœ‰å‹•ç‰©**éƒ½æœ‰ï¼šåå­—ã€å¹´é½¡ã€æœƒç™¼å‡ºè²éŸ³
- **ç‹—**æ˜¯ä¸€ç¨®å‹•ç‰©ï¼Œä½†æœƒã€Œæ±ªæ±ªå«ã€
- **è²“**ä¹Ÿæ˜¯ä¸€ç¨®å‹•ç‰©ï¼Œä½†æœƒã€Œå–µå–µå«ã€

**ç¹¼æ‰¿çš„å¥½è™•**ï¼š

1. **é¿å…é‡è¤‡å¯« code**ï¼ˆnameã€age ä¸ç”¨å¯«å…©æ¬¡ï¼‰
2. **å»ºç«‹éšå±¤é—œä¿‚**ï¼ˆDog æ˜¯ä¸€ç¨® Animalï¼‰
3. **å…±ç”¨åŸºæœ¬åŠŸèƒ½**ï¼Œä½†å¯ä»¥**å®¢è£½åŒ–ç‰¹æ®Šè¡Œç‚º**

**è¡“èª**ï¼š
- **çˆ¶é¡åˆ¥ï¼ˆParent Classï¼‰** = Animalï¼ˆåŸºç¤é¡åˆ¥ï¼‰
- **å­é¡åˆ¥ï¼ˆChild Classï¼‰** = Dogã€Catï¼ˆç¹¼æ‰¿è‡ª Animalï¼‰

````python
# çˆ¶é¡åˆ¥ï¼šå‹•ç‰©ï¼ˆå®šç¾©æ‰€æœ‰å‹•ç‰©å…±åŒçš„ç‰¹æ€§ï¼‰
class Animal:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    def eat(self):
        print(f"ğŸ– {self.name} æ­£åœ¨åƒé£¯...")
    
    def sleep(self):
        print(f"ğŸ˜´ {self.name} æ­£åœ¨ç¡è¦º...")

# å­é¡åˆ¥ï¼šç‹—ï¼ˆç¹¼æ‰¿ Animal çš„æ‰€æœ‰åŠŸèƒ½ï¼‰
class Dog(Animal):  # Dog ç¹¼æ‰¿ Animal
    def speak(self):
        print(f"ğŸ• {self.name} èªªï¼šæ±ªæ±ªæ±ªï¼")
    
    def fetch(self):
        print(f"ğŸ¾ {self.name} å»æ’¿çƒäº†ï¼")

# å­é¡åˆ¥ï¼šè²“ï¼ˆç¹¼æ‰¿ Animal çš„æ‰€æœ‰åŠŸèƒ½ï¼‰
class Cat(Animal):  # Cat ç¹¼æ‰¿ Animal
    def speak(self):
        print(f"ğŸ± {self.name} èªªï¼šå–µå–µå–µï½")
    
    def scratch(self):
        print(f"ğŸª‘ {self.name} æ­£åœ¨æŠ“æ²™ç™¼...")

# å»ºç«‹ç‰©ä»¶
dog = Dog("é˜¿ç¦", 3)
cat = Cat("å°èŠ±", 2)

print("--- ç‹—çš„è¡Œç‚º ---")
dog.eat()     # ç¹¼æ‰¿è‡ª Animal
dog.sleep()   # ç¹¼æ‰¿è‡ª Animal
dog.speak()   # Dog è‡ªå·±çš„æ–¹æ³•
dog.fetch()   # Dog è‡ªå·±çš„æ–¹æ³•

print("\n--- è²“çš„è¡Œç‚º ---")
cat.eat()     # ç¹¼æ‰¿è‡ª Animal
cat.sleep()   # ç¹¼æ‰¿è‡ª Animal
cat.speak()   # Cat è‡ªå·±çš„æ–¹æ³•
cat.scratch() # Cat è‡ªå·±çš„æ–¹æ³•

print("\nâœ¨ Dog å’Œ Cat éƒ½ç¹¼æ‰¿äº† Animal çš„ eat() å’Œ sleep()ï¼")
````

## 6ï¸âƒ£ å¤šå‹ï¼ˆPolymorphismï¼‰

### ğŸ­ ç”¨ã€Œèªªè©±æ–¹å¼ã€ä¾†ç†è§£

**å¤šå‹**çš„æ„æ€æ˜¯ï¼šã€ŒåŒæ¨£çš„å‹•ä½œï¼Œä¸åŒçš„çµæœã€ã€‚

ä¾‹å¦‚ï¼š
- ä½ å°**ç‹—**èªªã€Œå«ä¸€è²ã€â†’ æ±ªæ±ªæ±ª
- ä½ å°**è²“**èªªã€Œå«ä¸€è²ã€â†’ å–µå–µå–µ
- ä½ å°**é´¨**èªªã€Œå«ä¸€è²ã€â†’ å‘±å‘±å‘±

**åŒæ¨£éƒ½æ˜¯ã€Œå«ä¸€è²ã€**ï¼Œä½†**æ¯ç¨®å‹•ç‰©çš„å¯¦ä½œä¸åŒ**ã€‚

**å¤šå‹çš„å¥½è™•**ï¼š

1. **çµ±ä¸€ä»‹é¢**ï¼ˆéƒ½ç”¨ `speak()` é€™å€‹åå­—ï¼‰
2. **éˆæ´»æ“´å……**ï¼ˆè¦åŠ æ–°å‹•ç‰©å¾ˆç°¡å–®ï¼‰
3. **ç¨‹å¼ç¢¼æ›´ä¹¾æ·¨**ï¼ˆä¸ç”¨å¯«ä¸€å † if-elseï¼‰

````python
# ç¹¼çºŒç”¨å‰›å‰›çš„ Animal, Dog, Cat

# å†åŠ ä¸€å€‹ Duck é¡åˆ¥
class Duck(Animal):
    def speak(self):
        print(f"ğŸ¦† {self.name} èªªï¼šå‘±å‘±å‘±ï½")

# å»ºç«‹ä¸€ç¾¤å‹•ç‰©
animals = [
    Dog("é˜¿ç¦", 3),
    Cat("å°èŠ±", 2),
    Duck("å”è€é´¨", 5)
]

print("=== è®“æ‰€æœ‰å‹•ç‰©è¬›è©± ===")
for animal in animals:
    # åŒæ¨£éƒ½å‘¼å« speak()ï¼Œä½†æ¯ç¨®å‹•ç‰©çš„å¯¦ä½œä¸åŒ
    animal.speak()

print("\nâœ¨ é€™å°±æ˜¯å¤šå‹ï¼")
print("   åŒæ¨£çš„æ–¹æ³•åç¨±ï¼ˆspeakï¼‰ï¼Œä¸åŒçš„å¯¦ä½œçµæœã€‚")
print("   æˆ‘å€‘ä¸éœ€è¦çŸ¥é“æ˜¯ä»€éº¼å‹•ç‰©ï¼Œåªè¦å‘¼å« speak() å°±å¥½ï¼")
````

## 7ï¸âƒ£ dataclass ç°¡ä»‹

### ğŸ“¦ ä»€éº¼æ™‚å€™ç”¨ dataclassï¼Ÿ

å¦‚æœä½ çš„ class **ä¸»è¦æ˜¯ç”¨ä¾†å„²å­˜è³‡æ–™**ï¼Œè€Œä¸æ˜¯åšè¤‡é›œçš„é‚è¼¯ï¼Œå¯ä»¥ç”¨ `dataclass` ç°¡åŒ–å¯«æ³•ã€‚

**å‚³çµ±å¯«æ³• vs dataclass**ï¼š

å‚³çµ±ï¼š
```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y
```

dataclassï¼š
```python
@dataclass
class Point:
    x: int
    y: int
```

**dataclass è‡ªå‹•å¹«ä½ åšçš„äº‹**ï¼š
- âœ… è‡ªå‹•ç”¢ç”Ÿ `__init__`
- âœ… è‡ªå‹•ç”¢ç”Ÿ `__repr__`ï¼ˆå¥½çœ‹çš„é¡¯ç¤ºæ ¼å¼ï¼‰
- âœ… è‡ªå‹•ç”¢ç”Ÿ `__eq__`ï¼ˆå¯ä»¥æ¯”è¼ƒå…©å€‹ç‰©ä»¶ï¼‰

**ä½•æ™‚ç”¨ï¼Ÿ**
- âœ… ä¸»è¦ç”¨ä¾†å„²å­˜è³‡æ–™ï¼ˆåƒè³‡æ–™åº«çš„ä¸€ç­†è¨˜éŒ„ï¼‰
- âœ… ä¸éœ€è¦è¤‡é›œçš„åˆå§‹åŒ–é‚è¼¯
- âŒ éœ€è¦å°è£ã€é©—è­‰é‚è¼¯ â†’ ç”¨å‚³çµ± class

````python
from dataclasses import dataclass

# ç”¨ dataclass å®šç¾©ä¸€å€‹åº§æ¨™é»
@dataclass
class Point:
    x: int
    y: int

# ç”¨ dataclass å®šç¾©ä¸€å€‹ç”¢å“
@dataclass
class Product:
    name: str
    price: float
    stock: int = 0  # å¯ä»¥è¨­å®šé è¨­å€¼

# å»ºç«‹ç‰©ä»¶ï¼ˆè¶…ç°¡æ½”ï¼ï¼‰
p1 = Point(10, 20)
p2 = Point(10, 20)

product = Product("iPhone", 30000, 50)

# è‡ªå‹•ç”¢ç”Ÿçš„å¥½çœ‹é¡¯ç¤º
print(f"é» 1ï¼š{p1}")
print(f"é» 2ï¼š{p2}")
print(f"ç”¢å“ï¼š{product}")

# è‡ªå‹•å¯ä»¥æ¯”è¼ƒ
print(f"\np1 == p2? {p1 == p2}")

print("\nâœ¨ æ³¨æ„çœ‹ï¼Œæˆ‘å€‘æ²’å¯« __init__ï¼Œä½†å®ƒè‡ªå‹•å¹«æˆ‘å€‘åšå¥½äº†ï¼")
````

## 8ï¸âƒ£ è¿·ä½ å°ˆæ¡ˆï¼šTodo å¾…è¾¦äº‹é …ç®¡ç†ç³»çµ±

### ğŸ¯ å°ˆæ¡ˆç›®æ¨™

å»ºç«‹ä¸€å€‹ç°¡å–®çš„ Todo ç³»çµ±ï¼ŒåŒ…å«ï¼š

2. **TodoList**ï¼šç®¡ç†å¤šå€‹å¾…è¾¦äº‹é …ï¼ˆæ–°å¢ã€å®Œæˆã€é¡¯ç¤ºï¼‰

**ç”¨åˆ°çš„ OOP æ¦‚å¿µ**ï¼š
- âœ… Class èˆ‡ Object

- âœ… dataclassï¼ˆç°¡åŒ– TodoItemï¼‰

````python
from dataclasses import dataclass
from datetime import datetime

# å–®ä¸€å¾…è¾¦äº‹é …
@dataclass
class TodoItem:
    title: str
    done: bool = False
    created_at: str = None
    
    def __post_init__(self):
        # dataclass çš„ç‰¹æ®Šæ–¹æ³•ï¼Œåœ¨ __init__ å¾ŒåŸ·è¡Œ
        if self.created_at is None:
            self.created_at = datetime.now().strftime("%Y-%m-%d %H:%M")
    
    def mark_done(self):
        self.done = True
    
    def mark_undone(self):
        self.done = False
    
    def __str__(self):
        status = "âœ…" if self.done else "â¬œ"
        return f"{status} {self.title} (å»ºç«‹æ–¼ {self.created_at})"

# æ¸¬è©¦ TodoItem
item = TodoItem("å­¸ç¿’ Python OOP")
print(item)
item.mark_done()
print(item)
````

````python
# Todo æ¸…å–®ç®¡ç†å™¨
class TodoList:
    def __init__(self, owner):
        self.owner = owner
        self._items = []  # ç”¨ list å„²å­˜æ‰€æœ‰ TodoItem
    
    def add(self, title):
        """æ–°å¢å¾…è¾¦äº‹é …"""
        item = TodoItem(title)
        self._items.append(item)
        print(f"â• å·²æ–°å¢ï¼š{title}")
    
    def complete(self, index):
        """å®ŒæˆæŒ‡å®šç·¨è™Ÿçš„äº‹é …"""
        if 0 <= index < len(self._items):
            self._items[index].mark_done()
            print(f"âœ… å·²å®Œæˆï¼š{self._items[index].title}")
        else:
            print("âš ï¸ ç·¨è™ŸéŒ¯èª¤ï¼")
    
    def show_all(self):
        """é¡¯ç¤ºæ‰€æœ‰äº‹é …"""
        print(f"\nğŸ“‹ {self.owner} çš„å¾…è¾¦æ¸…å–®")
        print("=" * 50)
        if not self._items:
            print("ï¼ˆç©ºç©ºå¦‚ä¹Ÿï½ï¼‰")
        else:
            for i, item in enumerate(self._items):
                print(f"{i}. {item}")
        print("=" * 50)
    
    def show_pending(self):
        """åªé¡¯ç¤ºæœªå®Œæˆçš„äº‹é …"""
        print(f"\nâ³ {self.owner} çš„å¾…è¾¦äº‹é …ï¼ˆæœªå®Œæˆï¼‰")
        print("=" * 50)
        pending = [item for item in self._items if not item.done]
        if not pending:
            print("ğŸ‰ å…¨éƒ¨å®Œæˆäº†ï¼")
        else:
            for item in pending:
                print(f"  {item}")
        print("=" * 50)
    
    @property
    def total(self):
        """ç¸½äº‹é …æ•¸"""
        return len(self._items)
    
    @property
    def completed_count(self):
        """å·²å®Œæˆæ•¸é‡"""
        return sum(1 for item in self._items if item.done)
    
    @property
    def pending_count(self):
        """å¾…å®Œæˆæ•¸é‡"""
        return self.total - self.completed_count
````

````python
# å¯¦éš›ä½¿ç”¨ Todo ç³»çµ±
my_todos = TodoList("å°æ˜")

# æ–°å¢å¾…è¾¦äº‹é …
my_todos.add("å­¸ç¿’ Python OOP")
my_todos.add("å¯«ä½œæ¥­")
my_todos.add("é‹å‹• 30 åˆ†é˜")
my_todos.add("é–±è®€ä¸€ç« æ›¸")

# é¡¯ç¤ºæ‰€æœ‰äº‹é …
my_todos.show_all()

# å®Œæˆå¹¾å€‹äº‹é …
print("\n--- é–‹å§‹å®Œæˆäº‹é … ---")
my_todos.complete(0)  # å®Œæˆã€Œå­¸ç¿’ Python OOPã€
my_todos.complete(2)  # å®Œæˆã€Œé‹å‹• 30 åˆ†é˜ã€

# å†æ¬¡é¡¯ç¤ºæ‰€æœ‰äº‹é …
my_todos.show_all()

# åªé¡¯ç¤ºå¾…è¾¦äº‹é …
my_todos.show_pending()

# é¡¯ç¤ºçµ±è¨ˆè³‡è¨Š
print(f"\nğŸ“Š çµ±è¨ˆè³‡è¨Š")
print(f"ç¸½å…±ï¼š{my_todos.total} é …")
print(f"å·²å®Œæˆï¼š{my_todos.completed_count} é …")
print(f"å¾…å®Œæˆï¼š{my_todos.pending_count} é …")
````

````python
# æœ€å¾Œä¸€å€‹ç·´ç¿’ï¼šå»ºç«‹ä½ è‡ªå·±çš„ç¬¬ä¸€å€‹ Classï¼
# è©¦è‘—å»ºç«‹ä¸€å€‹ã€Œå­¸ç”Ÿã€é¡åˆ¥ï¼ŒåŒ…å«ï¼š
# - å±¬æ€§ï¼šå§“åã€å¹´ç´šã€æˆç¸¾åˆ—è¡¨
# - æ–¹æ³•ï¼šæ–°å¢æˆç¸¾ã€è¨ˆç®—å¹³å‡åˆ†æ•¸ã€é¡¯ç¤ºå­¸ç”Ÿè³‡è¨Š

class Student:
    def __init__(self, name, grade):
        self.name = name
        self.grade = grade
        self._scores = []
    
    def add_score(self, subject, score):
        self._scores.append({"ç§‘ç›®": subject, "åˆ†æ•¸": score})
        print(f"âœ… å·²æ–°å¢ {subject} æˆç¸¾ï¼š{score} åˆ†")
    
    def get_average(self):
        if not self._scores:
            return 0
        total = sum(item["åˆ†æ•¸"] for item in self._scores)
        return round(total / len(self._scores), 2)
    
    def show_info(self):
        print(f"\nğŸ‘¨â€ğŸ“ å­¸ç”Ÿè³‡è¨Š")
        print(f"å§“åï¼š{self.name}")
        print(f"å¹´ç´šï¼š{self.grade}")
        print(f"æˆç¸¾è¨˜éŒ„ï¼š")
        for item in self._scores:
            print(f"  - {item['ç§‘ç›®']}ï¼š{item['åˆ†æ•¸']} åˆ†")
        print(f"å¹³å‡åˆ†æ•¸ï¼š{self.get_average()} åˆ†")

# æ¸¬è©¦ä½ çš„ Student é¡åˆ¥
student = Student("å°æ˜", "é«˜ä¸€")
student.add_score("æ•¸å­¸", 85)
student.add_score("è‹±æ–‡", 90)
student.add_score("ç‰©ç†", 78)
student.show_info()

print("\nğŸ‰ æ­å–œï¼ä½ å·²ç¶“å®Œæˆ Python OOP å…¥é–€èª²ç¨‹ï¼")
print("ğŸ’ª ç¹¼çºŒç·´ç¿’ï¼Œä½ æœƒè¶Šä¾†è¶Šå²å®³çš„ï¼")
````
