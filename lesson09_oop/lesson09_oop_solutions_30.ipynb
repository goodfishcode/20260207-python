````python
class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author
        self.available = True
    
    def borrow(self):
        self.available = False
    
    def return_book(self):
        self.available = True

class Library:
    def __init__(self):
        self.books = []
    
    def add_book(self, book):
        self.books.append(book)
    
    def available_count(self):
        return sum(1 for book in self.books if book.available)

# æ¸¬è©¦
library = Library()
book1 = Book("Python å…¥é–€", "ä½œè€… A")
book2 = Book("JavaScript å…¥é–€", "ä½œè€… B")
library.add_book(book1)
library.add_book(book2)
assert_equal(library.available_count(), 2, "åˆå§‹å…©æœ¬å¯å€Ÿ")
book1.borrow()
assert_equal(library.available_count(), 1, "å€Ÿèµ°ä¸€æœ¬å¾Œ")
book1.return_book()
assert_equal(library.available_count(), 2, "æ­¸é‚„å¾Œ")
````

## Solution 29 â€” ç¶œåˆæ‡‰ç”¨ â€” æˆç¸¾ç®¡ç†ç³»çµ±

````python
class Student:
    def __init__(self, name):
        self.name = name
        self._scores = []
    
    def add_score(self, score):
        self._scores.append(score)
    
    @property
    def average(self):
        if not self._scores:
            return 0
        return sum(self._scores) / len(self._scores)

class Classroom:
    def __init__(self):
        self.students = []
    
    def add_student(self, student):
        self.students.append(student)
    
    def class_average(self):
        if not self.students:
            return 0
        total = sum(student.average for student in self.students)
        return total / len(self.students)

# æ¸¬è©¦
classroom = Classroom()
s1 = Student("å°æ˜")
s1.add_score(80)
s1.add_score(90)
s2 = Student("å°è¯")
s2.add_score(70)
s2.add_score(80)
classroom.add_student(s1)
classroom.add_student(s2)
assert_equal(s1.average, 85.0, "å°æ˜å¹³å‡")
assert_equal(s2.average, 75.0, "å°è¯å¹³å‡")
assert_equal(classroom.class_average(), 80.0, "å…¨ç­å¹³å‡")
````

## Solution 30 â€” çµ‚æ¥µæŒ‘æˆ° â€” éŠæˆ²è§’è‰²ç³»çµ±
