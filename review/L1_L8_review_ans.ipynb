### Q27ï¼šéè¿´ â€” éšä¹˜

ç”¨éè¿´è¨ˆç®— n çš„éšä¹˜ï¼ˆn!ï¼‰ã€‚

**æç¤º**ï¼š
- `factorial(0) = 1`
- `factorial(n) = n * factorial(n-1)`

**æ¸¬è©¦**ï¼š`factorial(5)` â†’ `120`

````python
# Q27 è§£ç­”
# å®šç¾©éè¿´å‡½å¼
def factorial(n):
    # åŸºç¤æƒ…æ³ï¼ˆçµ‚æ­¢æ¢ä»¶ï¼‰
    if n == 0 or n == 1:
        return 1
    # éè¿´æƒ…æ³
    return n * factorial(n - 1)

# æ¸¬è©¦
print(factorial(5))  # è¼¸å‡º 120
````

**è§£èªª**ï¼š
- éè¿´ï¼šå‡½å¼å‘¼å«è‡ªå·±

- åŸ·è¡Œéç¨‹ï¼š

  ```
  factorial(5)
  = 5 * factorial(4)
  = 5 * 4 * factorial(3)

  = 5 * 4 * 3 * 2 * 1
  = 120
  ```
