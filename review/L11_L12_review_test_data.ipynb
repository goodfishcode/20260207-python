````python
# üë®‚Äçüíª ‰ΩúÁ≠îÂçÄ
# Ë´ãÂú®ÈÄôË£°ÂØ´‰Ω†ÁöÑÁ®ãÂºèÁ¢º
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['Microsoft JhengHei']
plt.rcParams['axes.unicode_minus'] = False
plt.figure(figsize = (10,6))
x = df1['order_date']
y = df1['quantity'] * df1['unit_price']
plt.plot(x,y)
plt.title('ÊØèÊó•ÁáüÊî∂Ë∂®Âã¢')
plt.xlabel('Êó•Êúü')
plt.ylabel('ÁáüÊî∂(ÂÖÉ')
plt.xticks(rotation = 45)
````

````python
# Ëß£Á≠îÂæå
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['Microsoft JhengHei']
plt.rcParams['axes.unicode_minus'] = False

df_clean = df1.copy()
df_clean['order_date'] = pd.to_datetime(df_clean['order_date'])

df_clean['total_amout'] = df_clean['quantity'] * df_clean['unit_price']

daily_revenue = df_clean.groupby('order_date')['total_amout'].sum().sort_index()

plt.figure(figsize=(12,6))
plt.plot(daily_revenue.index, daily_revenue.values, marker='o', linewidth=2, markersize=6)

plt.title('ÊØèÊó•ÁáüÊî∂Ë∂®Âã¢')
plt.xlabel('Êó•Êúü')
plt.ylabel('ÁáüÊî∂(ÂÖÉ')

# Á∂≤Ê†º
plt.grid(True, alpha=0.3)
# ÊääÂúñÁâáÂç∞Âá∫‰æÜ
plt.show()
````
