## é¡Œç›® 14ï¼šfor è¿´åœˆ + write() - è½‰æ›ä¸¦å¯«å…¥æ–°æª”

## é¡Œç›® 19ï¼šè®€å–ä¸¦å»ºç«‹ç´¢å¼•åˆ—è¡¨

**é¡Œç›®ï¼š** è®€å– `courses.txt`ï¼Œç‚ºæ¯è¡ŒåŠ ä¸Šç·¨è™Ÿï¼Œå¯«å…¥ `courses_indexed.txt`ã€‚

**è§£æ³•æ€è·¯ï¼š**
1. è®€å– courses.txt çš„æ‰€æœ‰è¡Œ
2. ä½¿ç”¨ `enumerate(lines, 1)` å–å¾—ç´¢å¼•ï¼ˆå¾ 1 é–‹å§‹ï¼‰å’Œå…§å®¹
3. æ ¼å¼åŒ–å¾Œå¯«å…¥æ–°æª”æ¡ˆï¼ˆç·¨è™Ÿ. èª²ç¨‹åç¨±ï¼‰

```python
# è®€å–ã€åŠ ä¸Šç·¨è™Ÿä¸¦å¯«å…¥æ–°æª”
f_in = open('courses.txt', 'r')
lines = f_in.readlines()
f_in.close()

f_out = open('courses_indexed.txt', 'w')

for index, line in enumerate(lines, 1):
    course_name = line.strip()  # å»é™¤æ›è¡Œç¬¦è™Ÿ
    f_out.write(f'{index}. {course_name}\n')

f_out.close()

print("courses_indexed.txt å»ºç«‹å®Œæˆï¼")

# é©—è­‰å…§å®¹
f = open('courses_indexed.txt', 'r')
print(f.read())
f.close()

# é æœŸè¼¸å‡ºï¼š
# 1. Python Programming
# 2. Web Development
# 3. Data Science
```

## é¡Œç›® 20ï¼šç¶œåˆæ‡‰ç”¨ - å­¸ç”Ÿæˆç¸¾è™•ç†

**é¡Œç›®ï¼š** å»ºç«‹ `grades.txt`ï¼ˆå§“å,åœ‹æ–‡,è‹±æ–‡,æ•¸å­¸ï¼‰ï¼Œè¨ˆç®—å¹³å‡æˆç¸¾ä¸¦å¯«å…¥ `grades_avg.txt`ã€‚

**è§£æ³•æ€è·¯ï¼š**
1. å»ºç«‹åŒ…å«å››ä½å­¸ç”Ÿæˆç¸¾çš„æª”æ¡ˆï¼ˆCSV æ ¼å¼ï¼‰
2. é€è¡Œè®€å–ï¼Œä½¿ç”¨ `.split(',')` åˆ†å‰²å§“åå’Œæˆç¸¾
3. å°‡æˆç¸¾è½‰ç‚ºæ•¸å­—ä¸¦è¨ˆç®—å¹³å‡å€¼ï¼ˆç¸½å’Œ / 3ï¼‰
4. æ ¼å¼åŒ–è¼¸å‡ºï¼ˆä¿ç•™å…©ä½å°æ•¸ï¼‰ä¸¦å¯«å…¥æ–°æª”æ¡ˆ

```python
# ç¬¬ä¸€æ­¥ï¼šå»ºç«‹ grades.txt
f = open('grades.txt', 'w')
f.write('Alice,85,90,88\n')
f.write('Bob,78,82,80\n')
f.write('Charlie,92,95,90\n')
f.write('Diana,88,85,87')
f.close()

print("grades.txt å»ºç«‹å®Œæˆï¼")
```

```python
# ç¬¬äºŒæ­¥ï¼šè¨ˆç®—å¹³å‡ä¸¦å¯«å…¥ grades_avg.txt
f_in = open('grades.txt', 'r')
lines = f_in.readlines()
f_in.close()

f_out = open('grades_avg.txt', 'w')

for line in lines:
    # åˆ†å‰²è³‡æ–™
    data = line.strip().split(',')
    name = data[0]
    score1 = int(data[1])
    score2 = int(data[2])
    score3 = int(data[3])
    
    # è¨ˆç®—å¹³å‡
    average = (score1 + score2 + score3) / 3
    
    # å¯«å…¥æª”æ¡ˆï¼ˆä¿ç•™å…©ä½å°æ•¸ï¼‰
    f_out.write(f'{name}: {average:.2f}\n')

f_out.close()

print("grades_avg.txt å»ºç«‹å®Œæˆï¼")

# é©—è­‰å…§å®¹
f = open('grades_avg.txt', 'r')
print(f.read())
f.close()

# é æœŸè¼¸å‡ºï¼š
# Alice: 87.67
# Bob: 80.00
# Charlie: 92.33
# Diana: 86.67
```
